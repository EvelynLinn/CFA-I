# Alternative Investments

## Overview of Alternative Investments: Returns, Risks, and Portfolio Context
Alternative investments are alternatives to long-only stocks, bonds, and cash. Approximately 25% of global assets under management are classified as alternatives.

Alternative investments usually have the following characteristics:
- Narrow manager specialization
- Low correlation with traditional investments
- Limited regulation, transparency, and historical data
- Unique legal and tax issues
- High fees
- High use of leverage
- Concentrated positions
- Restrictions on redemptions

Alternative assets offer potentially high returns because managers are often able to profit from mispricings, illiquidity, and high leverage. (Although return correlations between alternative and traditional assets are relatively low during normal market conditions, they can increase sharply during periods of market turmoil.)

#### Categories of Alternative Investments
The main types of alternative investments that will be covered in this reading include:
1. Hedge funds
2. Private equity funds (many involve leveraged buyouts)
3. Real estate
4. Commodities
5. Infrastructure (for example, via public–private partnership (PPP) approach)
6. Other: tangible assets (such as fine wine, art, antique furniture and automobiles, stamps, coins, and other collectibles) and intangible assets (such as patents and litigation actions)

As measured by assets under management, commercial real estate is by far the largest class of alternative investments.
### Investment Structures
Many alternative asset funds are structured as `partnerships` with investors as `limited partners`. The asset manager serves as the `general partner`, earning a management fee based on the size of the fund and an incentive fee based on returns.

`High-water marks`, which reflects the highest cumulative return used to calculate an incentive fee, may be used so investors do not pay twice for the same performance gains. 

    For example, if a fund's value is initially 100 and rises to 120 in Year 1, incentive fees will be paid on that gain and 120 will be established as the high-water mark. If the fund's value falls back to 100 in Year 2 and rises to 120 again in Year 3, the high-water mark provision will ensure that the manager does not receive incentive fees for Year 3 returns that simply get investors back to where they were at the end of Year 1.

Because these funds are typically only available to qualified investors, they are subject to less regulation than the mutual funds and ETFs that are marketed to retail investors.
### Returns and Risks
While there are investment vehicles that offer passive exposure to alternatives (e.g., real estate, commodities, infrastructure), managers tend to pursue active returns. 

However, it is important to consider the additional risks associated with alternatives, which include:
- low liquidity, transparency, limited redemption availabilit, and the challenges of diversifying among managers. 
- Because leverage is more commonly used by alternative asset managers, returns on these investments are more sensitive to market conditions.

#### Common Risk Ratios
Measuring risk–return trade-offs.
##### Sharpe Ratio
Standard deviation and, by extension, the `Sharpe ratio`, are often inappropriate for assessing alternatives because these measures assume that returns are normally distributed and punish upside volatility. The ratio equally penalizes upside and downside volatility, fails to capture non-symmetric distributions, and may not fully reflect tails in return distributions. The return distributions to alternative investments are typically non-symmetric and skewed, making the Sharpe ratio a less-than-ideal measure. 
$$\frac{return- riskfree\ rate}{return\ standard\ deviation}$$
##### Sortina Ratio
Analysts tend to use the `Sortino ratio`, which emphasizes downside risk, has downside deviation rather than standard deviation as its denominator. 
$$\frac{mean\ realized\ return- target\ return}{standard\ deviation\ of\ the\ returns\ below\ the\ target\ return}$$

Note that using either of these measures to assess a hedge fund's performance fails to consider the correlation between its returns and returns on traditional assets, a less-than-perfect correlation between investments reduces the standard deviation of a diversified portfolio below the weighted average of the standard deviations of the investments. Other downside risk measures, such as the chance of losing a certain amount of money in a given period, are also used in practice. (`downside frequencies`)
### Portfolio Context: Integration of Alternative Investments with Traditional Investments
Investors can improve risk-adjusted returns for a portfolio of traditional assets by adding alternative investments, which offer diversification benefits. However, as noted, correlations between these traditional and alternative assets can be expected to increase during crises.

*Note: Why Investors Should Treat Alternative Investment Historical Records Critically*

Rreported return data can subject to biases. Historical return data may understate volatility for assets such as real estate and private equity, which often rely on appraisals (which are updated relatively infrequently) rather than market valuations. Indexes, particularly for hedge funds can have survivorship bias (excluding the impact of funds that have failed) and backfill bias (selectively including historical returns).
- Reported returns and standard deviations are averages and may not be representative of sub-periods within the reported period or future periods. 
- Many investments, such as direct real estate and private equity, are often valued using estimated (appraised) values rather than actual market prices for the subject investments. As a result, the volatility of their returns, as well as the correlation of their returns with the returns of traditional asset classes, will tend to be underestimated. 
- The record of alternative investment universes, such as hedge fund indexes, may be subject to a variety of biases, including survivorship and backfill biases. 
- Different weightings and constituents in index construction can significantly affect the indexes and their results and comparability. For example, commodity indexes can be weighted heavily in one particular sector, such as oil and gas.

## Hedge Funds
Historically, the first headge fund followed three key tenets: (1) Always maintain short positions, (2) always use leverage, and (3) charge only an incentive fee of 20% of profits with no fixed fees.

Common characteristics of hedge funds include:
- Aggressively managed with high return objectives
- Few restrictions on permissible investments
- Relatively high use of leverage and derivatives
- Willingness to take long and short positions
- Structured as private partnerships with relatively few investors
- Restrictions on redemptions (lockup period and notice period, fee charged to redeem shares)

`Funds of funds (FoFs)` are hedge funds that invest in a portfolio of other hedge funds. These investment vehicles offer diversification benefits and FoF managers may be able to negotiate better terms than those offered to individual investors. In exchange for this, FoFs charge an additional layer of fees. For `retail investors`, since a fund of hedge funds is diversified into multiple hedge funds, it has the disadvantage of offering lower potential return than a single hedge fund. A fund of hedge funds, though, has the advantages of being more accessible to retail investors and often being able to negotiate better redemption terms.
### Hedge Strategies
#### Equity Hedge Strategies
Equity hedge strategies take long and short positions in traditional publicly traded equities. This is a `bottom-up strategy`. Examples include:
- **Market neutral**: Long undervalued equities and short overvalued equities, but minimal net exposure to equity market risk (i.e., beta)
- **Fundamental growth**: Long positions in companies with fundamentals that indicate high growth potential
- **Fundamental value**: Long positions in undervalued companies, "deep value" managers may be willing to hold distressed securities
- **Quantitative directional**: Long and short positions based on technical analysis, some net long or short exposure depending on the expected direction of the overall equity market
- **Short bias**: Short positions in overvalued companies, possibly some long exposure if the overall market outlook is relatively bullish
- **Sector-specific**: Positions based on fundamental or technical expertise in a particular sector

#### Event-Driven Strategies
Event-driven strategies seek to profit from short-term events. Examples of this `bottom-up approach` include:
- **Merger arbitrage**: Typically long the target firm and short the acquirer, risk of losses if mergers are not completed
- **Distressed/restructuring**: Willing to hold the securities of companies that are on the verge of (or already in) the bankruptcy process with the expectation of profiting from restructurings
- **Activist**: Activist shareholders, often hedge funds, accumulate sufficient voting rights to influence a company's policies, board membership, and strategic direction
- **Special situations**: Pursuing opportunities other than mergers and restructuring, such as spinoffs or share repurchases

#### Relative Value Strategies
Relative value strategies seek to profit from `pricing discrepancies` by trading related securities, such as common and preferred stocks from the **same issuer**.
- **Fixed-income convertible arbitrage**: Executed by taking a long position in a company's convertible debt and a short position in its common stock, so any mispricings between these securities can be exploited to earn arbitrage profits
- **Fixed-income asset-backed**: Pursuing mispricings in asset-backed and mortgage-backed securities
- **Fixed-income general**: Taking positions in a variety of fixed-income instruments in pursuit of arbitrage profits
- **Volatility**: Options positions are used to express directional views on market volatility
- **Multistrategy**: Willing to pursue relative value arbitrage opportunities across a variety of asset classes and instruments

#### Macro Strategies
Macro strategies use a `top-down approach`, taking long and short positions based on the analysis of economic variables.
### Hedge Funds and Diversification Benefits
Positive hedge fund performance does not seem to persist for particular strategies. However, hedge fund managers are free to change strategies. Recent studies seem to indicate the diversification benefits are now less.

### Fees and Returns
Hedge funds tend to earn much higher fees. A common fee structure is “2 and 20”, which is 2% of the assets under management plus a 20% incentive fee. The incentive fee could be based on profit with or without management fees. Investors can reduce their fees by agreeing to longer lockup periods during which they are unable to withdraw their funds.

Performance fees may be contingent on the `hurdle rate` being achieved. If a `soft` hurdle rate is used, the performance fee is paid on the full return as long as the hurdle rate is exceeded. With a `hard` hurdle rate, the incentive fee is only paid on the portion of the return in excess of the hurdle rate.

A `high-water mark` provision ensures that performance fees will not be paid on any returns that represent a recovery from a previous high. Fees may be lower if the investor agrees to keep the money in the fund for a longer period of time.

Many hedge funds do not remain in business long, which can lead to survivorship bias. `Survivorship bias` results from only including current investment funds in the indexes. Thus, funds that have failed are not included in the index. This means that the return measured in the index is artificially high.

`Backfill bias` results from including previous return data for funds that enter the index. This also results in artificially high returns measured in the index. `Sample selection bias` is observed in transaction-based real estate indexes.
### Leverage and Redemption
Hedge funds use high degrees of leverage, which magnifies both gains and losses. Leverage can be achieved by borrowing through a prime broker and/or using derivatives (e.g., futures, options). When derivatives are used, hedge funds must post collateral, which acts as a drag on returns.

Hedge fund losses can be magnified if investors are making withdrawals during periods of poor performance. A large number of investors redeeming shares causes a forced liquidation of the fund, which results in an even bigger drawdown. The liquidation causes a magnification of the losses, leading to redemption requests from the remaining investors, thus fully liquidating and closing the fund. Managers seek to prevent this by requiring lockup periods.
### Hedge Fund Valuation Issues
Hedge funds can be valued from daily to quarterly with `market values` or `estimated prices`. Some positions are very illiquid, which necessitates estimated prices. Valuations should be consistent for each period. A liquidity discount may need to be applied.
### Due Diligence for Investing in Hedge Funds
For both individual hedge funds and FoFs, investors must perform sufficient due diligence. Factors that should be considered in this process include:
- Investment strategy and process
- Competitive advantage
- Track record (make sure the calculations and fee structure are understood)
- Size and longevity (older funds probably have satisfied past investors)
- Management style
- Key-person risk
- Reputation
- Investor relations
- Plans for growth
- Systems risk management (make sure the process is rigorous)

## Private Equity
Private equity generally entails investing in private companies. The level of activity in private equity is typically cyclical. Categories of private equity include 
- `leveraged buyouts`
- `venture capital`
- `development capital`
- `distressed investing`

### Private Equity Structure and Fees
Like partnerships, private equity firms are the general partner with investors being the limited partners. Most private equity firms charge both management fees (1% to 3% of committed capital) and incentive fees. Committed capital may not all be invested initially. The incentive fee is normally 20% of the profit. No incentive fees are normally paid until the limited partners have received the principal back.

A private equity fund's limited partners can protect themselves against the possibility of earning a negative return on their initial investment by requiring a `clawback provision`, which requires a fund's general partner to return any incentive fees to the limited partners until a specified threshold has been met (e.g., investors have had the value of their initial investment returned).
### Private Equity Strategies
#### Leveraged Buyouts
Leveraged buyouts (`LBOs`) generally take a public company private. The buyout could be by existing management (i.e., `management buyout`) or by new management (i.e., `management buy-in`). LBOs seek to add value by improving operations.

**Debt is used in buyouts to leverage the equity returns**. `Leveraged loans` (usually from banks) carry covenants that protect the investors. They are often senior to bond debt, which is also used in the financing.

`Mezzanine financing` can also be used, which is debt or preferred shares that may be converted into common stock. 
- High yield debt is commonly included in the capital structure of LBOs. 
- As an alternative, convertible debt offers greater potential upside because it can be exchanged for common stock. However, convertible debt is less senior to high yield debt and the issuer is typically required to pay a higher coupon rate.

Characteristics that make a company an attractive takeover target by an LBO include:
- Undervalued stock price
- Willing management
- Inefficient companies
- Strong and sustainable cash flow
- Low leverage
- Significant physical assets

#### Venture Capital (VC)
Venture capital can be used for a company at inception or one that is getting ready to go public. Investors in the early stages expect higher returns. Venture capitalists are active investors.
1.  `Formative-stage financing` includes angel investing (idea stage), seed-stage financing, and early-stage financing. It is usually done through ordinary or convertible preferred shares. Management still keeps control.
  - `Angel investments`, which are typically made by **individuals**, support activities such as the development of a business plan.
  - The `seed stage` supports market research and product development and is generally the first stage at which venture capital funds invest. The seed stage follows the angel investing stage. 
  - `Early-stage` investments typically fund the initial phase of commercial production.

2. Later-stage financing is provided after sales but before any IPO. Management sells control of the company.
3.  `Mezzanine-stage financing` prepares a company for the IPO. It is provided after a company has started commercial production and is used to bridge the gap in advance of an IPO.

#### Other Private Equity Strategies
`Development capital` funds business growth. Often management is selling a minority interest of shares.
#### Exit Strategies
Private equity firms usually only hold positions for about five years. There are several common exit strategies, including:
- `Trade sale` involve selling to a strategic buyer like a competitor. This can provide immediate cash at a high valuation with little public knowledge. Because strategic buyers are often competitors within the same industry, the portfolio company's managers are more likely to oppose a private equity firm's efforts to execute such a transaction. This does limit the pool of buyers and may be less valuable than an IPO.
- `IPOs` sell the company to the public. This could be advantageous because it will likely get a high price, management is retained, and has potential to retain some upside. There is a high transaction cost fee, and long lead times are required with such disclosure.
- In `recapitalization`, the private equity firm still retains control and is able to extract some money from the deal. Private equity firms use recapitalizations as a complement to other exit strategies. For example, if interest rates fall, the firm can recapitalize, which allows it to earn a partial return immediately while retaining both control of the portfolio company and the ability to exit later via a trade sale or an initial public offering.
- `Secondary sales` can be made to other private equity firms or investors. When a private equity firm sells its position in a portfolio company to another private equity firm, the transaction is classified as a secondary sale.
- `Liquidations` occur when things go poorly. Portfolio companies, or certain assets, are sold for the purpose of salvaging any value.

### Private Equity: Diversification Benefits, Performance, and Risk
Private equity funds can get higher returns through influence over private operations and the use of leverage. They can also provide diversity to portfolios of publicly traded securities.

Private equity return indexes rely on `self-reporting`, so it is subject to survivorship, backfill, and other biases. Also, prices are updated less frequently which makes the volatility appear low.

Studies have shown the top private equity managers are much better than the average.
### Portfolio Company Valuation
- Market or Comparable<br>
A company is valued by using multiples of a different measure (e.g., EBITDA, net income).
- Discounted Cash Flow (DCF)<br>
A company is valued by using the present value of expected future cash flows. The weighted average cost of capital can be used as the discount rate.
- Asset-based<br>
A company is valued based on the underlying asset values. Analysts could use fair values or liquidation values.
### Private Equity: Investment Considerations and Due Diligence
Current and anticipated economic conditions are very important in the analysis. Patience is needed for long-term commitments and illiquidity.

## Real Estate
Real estate investing offers the following:
- Competitive returns from income and capital appreciation
- Multiple-year leases that insulate from economic cycles
- Diversification with other asset classes
- Inflation hedge from adjustable rents

Real estate ownership can include 
- air rights
- mineral rights
- surface rights

Real estate can be residential or commercial. It has features that differ from other asset classes, including 
- indivisibility
- uniqueness
- fixed location

### Forms of Real Estate Investment
Real estate can be classified as `debt` or `equity` and as `private` or `public`. Equity investments are direct investments. `Real estate investment trusts (REITs)` sell shares to raise money to purchase properties.

- Direct ownership can be free and clear or come with a `lien` from mortgage financing, which is usually a stand-alone mortgage loan. 
- `Pooled vehicles` such as `real estate limited partnerships` and REITs can be used. 
- `Securitization of mortgages` provides institutional investors with diversified fixed income securities.

### Real Estate Investment Categories
#### Residential Property
Many individuals directly own property that they occupy. Often borrowed funds are used to help with the purchase. Lenders must do due diligence (such as credit of borrower or value of the property). Securitization allows for indirect debt investment.
#### Commercial Real Estate
Commercial property is appropriate for investors with long-term horizons and little liquidity needs. Commercial property requires experienced professional management. The lender must consider the creditworthiness of the borrower and the ability of the property to generate sufficient revenue. The loan-to-value ratio is critical.
#### REIT Investing
REITs can either be mortgage REITs or equity REITs. A majority of the income must be distributed to shareholders.
#### Mortgage-Backed Securities (MBS)
Mortgages can be pooled and then divided into classes (or tranches). The different classes can have different degrees of risk.
#### Timberland and Farmland
Historically, these assets have not been highly correlated with other asset classes. They also offer hedges against inflation. The key return drivers are harvest quantities, commodity prices, and land prices. Timberland is more flexible with respect to harvesting, since timber can be stored easily by not harvesting, while farmland must be harvested when ripe.
### Real Estate Performance and Diversification Benefits
A real estate index can be an `appraisal index`, `repeat sales index`, or a `REIT index`.
1. Appraisal values are subjectively determined by experts for the appraisal index. The infrequency of appraisals may understate volatility.
2. Repeat sales indexes use **transaction prices** to update the values. This is a sampling approach since only a few numbers of properties are sold each period. These indexes are affected by sample selection bias.
3. REIT indexes use the prices of publicly traded REITs.

### Real Estate Valuation
Real estate values must be estimated until actually sold. Appraisals can be done with several methods, including:
- Comparable sales approach – approximate value based on sales of similar properties
- Income approach – direct capitalization or discounted cash flow
  - Direct capitalization estimates the property value from the level and quality of net operating income (NOI). The NOI is divided by the capitalization rate to get the estimated value.
  - The discounted cash flow projects and discounts the cash flows. Usually, a sale value is estimated at the end of the projection.
- Cost approach – replacement cost of land and building

#### REIT Valuations
REIT valuations depend on the underlying pool of securities. The valuation can be either income-based or asset-based.

The `income-based approach` is similar to the direct capitalization approach. Either the `funds from operations (FFO)` or the `adjusted funds from operations (AFFO)` can be used. The AFFO adjusts for recurring capital charges.

`Asset-based approaches` calculate the underlying net asset values of the REIT.
### Real Estate Investment Risks
Real estate investments are subject to economic conditions and interest rates. Returns are also impacted by the management of the properties. 

Leverage magnifies gains and losses.

## Commodities
Returns for commodities (physical products) are based on price changes. Transportation and storage costs lead many investors to trade commodity derivatives. It is important to understand the producers and users of commodities before investing.

Commodities include 
- energy
- base and precious metals
- agriculture. 

Classifications are also done by location and quality.

Physical trading of commodities is typically limited to parties that are directly involved in its supply chain. For example, farmers and food producers are actively involved in the physical trading of wheat. If a commodity is physically traded by parties outside of its supply chain, it is likely `non-perishable`, has a high value relative to its weight and volume, and capable of being stored at a relatively low cost. Gold is an example of a commodity with these characteristics.

Many invest in commodities to hedge against inflation.
### Commodity Derivatives and Indexes
Commodity derivatives include futures, forwards, options, and swaps. Forwards and futures are obligations to transact at a later date. Forwards carry more counterparty risk because they are traded over the counter.

Options contracts give the holder the right to transact in the future but not the obligation. 

Swaps are agreements to exchange multiple cash flows in the future.

**Commodity indexes are usually based on the futures price rather than the spot price of the commodity.**

### Other Commodity Investment Vehicles
It is most common to directly invest in commodities or commodity derivatives. There are other means, such as:
- Exchange-traded funds (ETFs)
- Common stock of a company exposed to a particular commodity
- Managed futures funds
- Individual managed accounts
- Specialized funds in commodity sectors

### Commodity Performance and Diversification Benefits
Commodity investments offer potential returns, portfolio diversification, and inflation protection. Historical returns have shown low correlations with other asset classes. Commodity prices do affect inflation calculations, so they should provide a natural hedge.

Returns on commodity investments are based on **changes in price alone** and do not include an income stream such as interest, dividends, or rent. 
### Commodity Prices and Investments
Commodity spot prices are a function of supply and demand.

Supply is dependent on production and inventory levels. Often supply cannot be changed quickly.

Demand is based on the needs of end-users. The economic environment can quickly change demand. The mismatch between supply and demand leads to price volatility.

#### Pricing of Commodity Futures Contracts
Commodity futures contracts are traded on exchanges. The buyer (long side) is obligated to take delivery of the commodity and pay the settlement price. The seller (short side) is obligated to deliver the commodity. Positions are marked to market daily to protect the exchange against default risk.

The futures price is approximated with:
$$Futures\ Price≈(Spot\ Price)(1+r)+Storage\ Costs−Convenience\ Yield$$
The above formula is necessary to preclude arbitrage.

Futures prices may be higher or lower than spot prices depending on the convenience yield. When futures prices are higher than the spot price, the commodity forward curve is upward sloping and the prices are referred to as being in contango. Contango occurs when there is little or no convenience yield. When futures prices are lower than the spot price, the commodity forward curve is downward sloping and the prices are referred to as being in backwardation. Backwardation occurs when the convenience yield is high.

- `Contango` occurs when the futures price is higher than the spot price. This happens when there is little or no convenience yield.
- `Backwardation` occurs when the futures price is less than the spot price due to a high convenience yield. The futures price then generally rolls up (moves up along the forward curve) to the spot price curve as the expiry date of the futures contract approaches, which results in a positive roll yield for the long positions.

        It is important to understand futures contracts and the sources of return for each commodity futures contract because commodity investments often involve the use of futures contracts. These contracts trade on exchanges. The buyer (i.e., the long side) of a futures contract is obligated to take delivery of the commodity or its cash equivalent based on the spot price at expiration and will pay a settlement price. The settlement price is calculated as specified in the contract based on its standard quantity (e.g., 1,000 barrels for Brent Crude Oil). In other words, the long side is obligated to buy the commodity at the final settlement price and take delivery at one of the locations specified in the contract. The long side of a futures contract increases in value when the underlying commodity increases in value. The seller of a futures contract (i.e., the short side) is obligated to deliver the commodity at the location specified in the contract or its cash equivalent based on the settlement price at expiration.

There are three sources of yield for commodity futures contracts:
1. `Roll yield`: With a high enough `convenience yield`, `backwardation` will occur. The futures price will naturally “roll up” to the spot price as the expiration date approaches.
2. `Collateral yield`: Typically, collateral will be invested in very safe securities.

## Infrastructure 
The assets underlying infrastructure investments are real, capital intensive, long-lived assets intended for public use. Most are financed by governments, although they now often have private financing. High entry barriers protect the investors and inelastic demand for assets and services typically provide investors with stable cash flows. These long-term investments also provide protection against inflation.
### Categories of Infrastructure Investments
Infrastructure investments can be categorized by the underlying assets, stage of development, or geographic location.
#### By Underlying
The underlying assets can be economic or social. `Economic infrastructure` assets supports economic activities, such assets include transportation (e.g., roads and bridges) and utility (e.g., power generation and waste removal) assets. `Social infrastructure` assets include educational and healthcare facilities.
#### By Stage
By stage of development, infrastructure assets can be `brownfield` (already existing) or `greenfield` (not yet constructed). 

Infrastructure investments are available globally, with risk and return varying by location.
### Forms of Infrastructure Investments
Infrastructure investments can either be direct or indirect. 
- `Direct investments` allow for more control but they require very large and illiquid investments. 
- `Indirect investments` are more common. This can be done through shares of companies, exchange-traded funds, or private equity funds. Publicly traded infrastructure securities provide more liquidity but represent a very small segment of the universe.
### Risk and Return Overview
Many infrastructure investments have a low risk with stable cash flows. Others are exposed to volatile cash flows, financing risk, operational risk, and construction risk. Regulatory risk is also common because essential services are being provided.

## Other Alternative Investments
Many other alternative investments are called `collectibles`. This could include anything from fine art to jewelry. Collectibles provide no income and are very illiquid. `Capital appreciation` (and `depreciation`) is the source of return. Storage costs can be significant. Several indexes strive to track collectible values.

Collectibles, including art, fine wine, rare coins, and stamps, provide a convenience yield -- investors can derive enjoyment from holding these items, even if they do not use them. However, investing in collectibles requires a high level of expertise in order to identify truly valuable items. Additionally, the prices of collectibles can be highly volatile.

## Risk Management Overview
Many risks must be considered, such as limited transparency, illiquidity, operational risk, and counterparty risk. Investment risk should be monitored by a chief risk officer who is separated from the investment process.
### Investment and Risk Management Process
The investor must decide how much to allocate to alternative investments. Risk must be considered and due diligence conducted. `Third-party custody` of assets and independent verification of results can help reduce the chance of an investor being defrauded. 

Risks vary by investments. Many alternative investments are illiquid with long horizons, so upfront due diligence is critical.

Historical returns may not accurately reflect the volatility of alternative investments. Performance can be highly correlated with the business cycle. Also, returns could vary greatly by managers. The benefits and risks of leverage should be understood. Performance fees could align the interest of managers and investors.

Due diligence is more critical for private investments than for public investments. Regular independent valuations of illiquid securities are needed. **Limits should be placed on leverage, security type, sector, and geography**. Exposure to counterparty risk should be monitored.
### Risk-Return Measures
The Sharpe ratio may not be appropriate for alternative investments because the standard deviation is not reliable due to illiquid investments. Returns tend to be smoothed when using estimated prices.

Also, many alternative investment returns are not normally distributed. Rather, they tend to be `leptokurtic` (fat tails) or `negatively skewed` (long left tails). The standard deviation measure assumes normally distributed returns. The normal distribution assumption can safely be made for the returns of long-only investments in traditional asset classes, but not for alternative investments.

The `standard deviation measure` accounts for much downside risk but underestimates it. Using the standard deviation measure in isolation to assess the performance of a single manager fails to reflect the benefits of portfolio diversification.

`Downside risk measures`, such as `value at risk (VaR)`, are more appropriate. VaR measures the minimum loss expected of a given time period at a given probability. Other measures include the `shortfall risk` and `Sortino ratio`. The Sortino ratio is able to reflect the risk of tail events by using downside deviation as a measure of risk. `Stress testing` is also very important.
### Due Diligence Overview
Management selection is critical. The compensation should align the interests of investors. Fraud must be avoided. From a due diligence perspective, diversifying a portfolio's allocation to alternative investments across multiple managers provides greater protection against possible fraud. This is because entrusting assets to just one manager increases an investor's exposure to potential fraud.
