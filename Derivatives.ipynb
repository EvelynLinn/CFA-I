# Derivative Markets and Instruments
`Derivatives`—financial instruments whose prices are derived from the value of some underlying asset (referred to as the `underlying`) —have become increasingly important for managing financial risk, exploiting investment opportunities, and creating synthetic asset class exposure. As in other security markets, arbitrage and market efficiency play a critical role in establishing prices for these securities. They have a defined and limited life.

## Introduction
Perhaps the distinction that best characterizes derivatives is that they usually transform the performance of the underlying asset before paying it out in the derivatives transaction.

Derivatives are similar to insurance in that both allow for the transfer of risk from one party to another. The buyer of a derivative contract has a `long position` and the seller has a `short position`.

*(Note: Investors maintain “long” security positions in the expectation that the stock will rise in value in the future. The opposite of a “long” position is a “short” position. A "short" position is generally the sale of a stock you do not own. Investors who sell short believe the price of the stock will decrease in value.)*

The so-called “underlying asset” is often simply referred to as the underlying, whose value is the **source of risk**. Common derivatives underlyings are:
- equities
- fixed-income securities 
- currencies
- commodities

Other derivatives underlyings include:
- interest rates
- credit
- energy
- weather
- other derivatives

Thus, like insurance, derivatives pay off on the basis of a source of risk, which is often, but not always, the value of an underlying asset. And like insurance, derivatives have a definite life span and expire on a specified date.
#### Classes (or Forms)
There are two general classes of derivatives:
1. Some provide the ability to lock in a price at which one might buy or sell the underlying. Because they force the two parties to transact in the future at a previously agreed-on price, these instruments are called `forward commitments`: 
  - forward contracts
  - futures contracts
  - swaps
2. `Contingent claims` provides the right but not the obligation to buy or sell the underlying at a pre-determined price (because the choice of buying or selling versus doing nothing depends on a particular random outcome, these derivatives are called contingent claims):
  - options

#### Purposes
Derivatives can be used to create strategies that would not exist in their absence. For example, derivatives make it easier to go short, thereby benefiting from a decline in the value of the underlying.

Leverage can also be easily created because investments are small relative to the value of the underlying. As such, small movements in the underlying can lead to fairly large movements in the amount of money made or lost on the derivative.

Derivatives usually have lower transaction costs than comparable spot market transactions and more liquidity compared to their underlying asset.  

They are critical to the risk management process because they can be used to adjust, reduce, or eliminate exposure to certain risks easily and quickly. *(For example, a shareholder of a company can reduce or even completely eliminate the market exposure by trading a derivative on the equity. Holders of fixed-income securities can use derivatives to reduce or completely eliminate interest rate risk, allowing them to focus on the credit risk. Alternatively, holders of fixed-income securities can reduce or eliminate the credit risk, focusing more on the interest rate risk.)*  


## Derivative Market Structure
Certain derivatives trade on `public exchanges` while others are traded in the `over-the-counter (OTC)` market.
### Exchange-Traded Derivatives Markets
Exchange-traded derivatives are standardized contracts, which means variations in terms such as expiration and underlying securities are limited. Standardized terms increase market liquidity because all investors are trading the same contract. `Market makers` and `speculators` are active participants on derivatives exchanges.

Clearing and settlement operations are smooth and come with a `credit guarantee`. The clearinghouse is able to guarantee the losing party's performance by requiring margin bonds of the participants. Exchange markets are also transparent. This information is useful to many but also results in a loss of privacy.
### Over-The-Counter Derivatives Markets
The OTC derivatives market operates through an informal network of dealers (usually banks) that buy and sell contracts at their discretion. OTC derivatives contracts are not standardized, which creates more risk for the dealer because it is difficult to find identical contracts to offset risk. Instead, dealers manage their risk exposures by taking offsetting positions in similar contracts.

OTC contracts are not universally illiquid. Some can have higher trading volumes than certain exchange-traded derivatives contracts. The OTC derivatives market has traditionally been lightly regulated, which gives participants more privacy and flexibility.

## Types of Derivatives
### Forward Commitments
Forward commitments can be exchange-traded contracts (e.g., futures) or over-the-counter (OTC) contracts (e.g., forwards and swaps).
#### Forward Cotracts
A forward contract is an agreement in advance to make a trade. One party agrees to pay the `forward price` for an underlying asset and the other party agrees to sell at that price on a specific date. *(The buyer is called the long party and the seller is called the short party.)*

Contract terms (e.g., where and when delivery will be made, the quantity of the underlying asset to be delivered) can be **customized** to meet the specific needs of each party. Forward contracts are regularly written on a wide variety of underlying assets.

Forward contracts have **zero value at inception**, so no money is exchanged at that time. They become more or less valuable to the respective parties as the price of the underlying asset fluctuates during the term of the contract. The value of a forward contract depends on the type of position:

To illustrate the payoff of a forward contract, start with the assumption that we are at time $t = 0$ and that the forward contract expires at a later date, time $t = T$. The `spot price` of the underlying asset at time 0 is $S_0$ and at time $T$ is $S_T$. The notation $F_0(T)$ denotes that this value is established at time $0$ and applies to a contract expiring at time $T$. $F_0(T)$ is the forward price. Roll forward to time $T$, when the price of the underlying is $S_T$. The long is obligated to pay $F_0(T)$, for which he receives an asset worth $S_T$. 

Long: $S_T−F_0(T)$<br>
Short: $−[S_T−F_0(T)]=F_0(T)−S_T$

As shown in the payoff formulas, the long party gains if the price of the underlying asset is above the forward price at expiry. Forward contracts are a zero-sum game because the long party's gain is equal to the short party's loss (or vice versa if the price of the underlying finishes below the forward price). The gain from owning the underlying would be $S_T ‒ S_0$, whereas the gain from owning the forward contract would be $S_T ‒ F_0(T)$. 

Each party faces the risk the other party will default, although only the "winning" party is exposed to default risk at any given time during the life of the contract. This usually limits participation to creditworthy entities. The forward market is private and largely unregulated.

Forward contracts can be used to hedge various risks. For example, manufacturers can use forward contracts to lock in the price of their raw materials. Many forward contracts require the seller to physically deliver the underlying asset to the long party. `Non-deliverable forwards (NDFs)` do not require physical delivery. Instead, the losing party makes a cash payment to the winning party for the amount of the difference between the forward price and the spot price of the underlying asset at expiry.
#### Futures
Futures contracts are specialized versions of forward contracts that have been standardized and that trade on a futures exchange. A futures contract is similar to a forward contract in that there is an agreement to purchase an underlying asset for a specific price on a specified date. Differences between forwards and futures include:
- Standardization: Futures contracts are `standardized`, while forward contracts are customized terms.
- Futures contracts trade on a `public exchange` with a `clearinghouse` that guarantees the performance of all traders. Forward contracts trade OTC with no performance guarantees, so traders must do their own assessment of counterparty credit risk.
- `Daily settlement`: Futures contracts are `marked to market`, which allows traders to realize gains or losses at the end of each day. By contrast, gains and losses for forward contracts are not realized until expiration. (Futures contracts are more liquid because it is much easier to take offsetting positions when trading standardized contracts. By contrast, forward contracts are usually held to maturity.)

The account is specifically referred to as a `margin account`. A `margin call` is made if an account balance drops below the `maintenance margin`. After a margin call, the party must deposit sufficient funds to bring the account balance back to the initial margin level. Note that it is not sufficient to bring the balance back to the maintenance margin.

    Note: (Margin account is used for an amount of money that a trader deposits in a margin account. The term is derived from the stock market practice in which an investor borrows a portion of the money required to purchase a certain amount of stock. In futures markets, there is no borrowing so the margin is more of a down payment or performance bond.)

Futures exchanges minimize their losses by requiring an `initial margin deposit` (typically less than 10% of the futures price) from both parties upon initiation. It serves as a performance bond to provide some assurance of the parties' ability to cover possible future losses. At the end of each trading day, gains are credited to one party's margin account based on the change in the futures price. An equivalent amount is debited from the balance of the other party's account.

**Price limits:** Some futures contracts have price limits that restrict the trading range/band. If prices reach these limits (limit up), trading will temporarily stop until a trade takes place above the `limit down` price or below the `limit up` price.

**Settlement:** Like forward contracts, futures may be settled in cash or with physical delivery of the underlying asset. Regardless of the settlement terms, **the futures price will `converge` to the spot price at expiration**.

    In the futures markets, the exchange specifies whether physical delivery or cash settlement applies. In physical delivery contracts, the short is required to deliver the underlying asset at a designated location and the long is required to pay for it. Delivery replaces the mark-to-market process on the final day. It also ensures an important principle that you will use later: The futures price converges to the spot price at expiration. Because the short delivers the actual asset and the long pays the current spot price for it, the futures price at expiration has to be the spot price at that time. Alternatively, a futures contract initiated right at the instant of expiration is effectively a spot transaction and, therefore, the futures price at expiration must equal the spot price. Following this logic, in cash settlement contracts, there is a final mark to market, with the futures price formally set to the spot price, thereby ensuring automatic convergence.

At any given time, the number of outstanding contracts is called the `open interest`. Each contract counted in the open interest has a long and a corresponding short. 

**Payoff:** The total payoff for a futures contract is the same as for a forward contract (see detailed deduction in textbook, which is straightforward), but the timing is different. Forward contracts realize the full amount at maturity, while futures contracts make and receive cash flows daily. (related: what's called an "offset".)

Futures prices of nearby expiring contracts are often used as proxies for spot prices, particularly in decentralized spot markets, such as gold, which trades in spot markets all over the world. (One of the ways how derivatives affect its underlying's values?)
#### Swaps
The concept of a swap is that two parties exchange (swap) a series of cash flows. One set of cash flows is variable or floating and will be determined by the movement of an underlying asset or rate. The other set of cash flows can be variable and determined by a different underlying asset or rate, or it can be fixed. 

A swap is a series of forward contracts, meaning that cash flows are exchanged on more than one date. Like forward contracts, swaps are private OTC transactions and each party is exposed to the risk that the other party will default. (Combinations of futures contracts expiring at different dates are often compared to swaps.)

They are often used by parties seeking to convert their `fixed-rate debt` into `floating-rate obligations` (or vice versa). This is called a `fixed-for-floating interest rate swap`, or a `plain vanilla swap`.

As with futures and forwards, no money changes hands at the start; thus, the value of a swap when initiated must be zero. The fixed rate on the swap is determined by a process that forces the value to zero, a procedure that will be covered later in the curriculum. As market conditions change, the value of a swap will deviate from zero, being positive to one party and negative to the other.

A swap has no net value at inception, so no money is exchanged when the contract is signed. Periodic payments are made based on the contract's `notional principal` and the rates that each party has agreed to pay. On each payment date, a net payment is made by the party owing more. (The swap has no principal, but it does have a balance of sorts, called the notional principal, which ordinarily matches the loan balance. A loan with only one principal payment, the final one, will be matched with a swap with a fixed notional principal. An amortizing loan, which has a declining principal balance, will be matched with a swap with a pre-specified declining notional principal that matches the loan balance.)

### Contingent Claims
A contingent claim is a derivative in which the outcome or payoff is dependent on the outcome or payoff of an underlying asset. For contingent claims, the payoff occurs if a specific event happens. Examples include options, credit derivatives, and asset-backed securities.
#### Options
An option gives one party the right, but not the obligation, to buy or sell an underlying asset. The terms of an option (e.g., the underlying asset, exercise price, date, quantity) are specified in advance.

A call option grants the right to buy the underlying asset and a put option grants the right to sell. The party that buys an option (either a call or a put) is called the long party, while the option seller is called the short party. For example, an investor who sells a put is said to have a short put position.

*Unlike forward commitments, a **cash payment** is made when the contract is signed. The right to buy or sell will only be exercised if doing so would be profitable for the party holding the option. If not, the option contract will be allowed to expire unused. The ability to profit from good outcomes and walk away from bad outcomes is valuable, so the option buyer is required to make an upfront payment (called a premium) to the option seller.*

The value of an option after initiation depends on the price of the underlying asset relative to the `exercise price`. For example, if a stock is trading for $40, a call option with an exercise price of \\$35 is described as `being in the money` because the call buyer has the right to buy the stock for less than its market value. The same option would be described as `being at the money` if the stock was trading at the \$35 exercise price or `out of the money` if it was trading below that level.

    The fixed price at which the underlying asset can be purchased is called the exercise price (also called the “strike price,” the “strike,” or the “striking price”). 

If $X$ is the exercise price and $c_0$ and $p_0$ are the premiums for call and put options, respectively, then the payoff and profit to the call and put buyers at maturity time $T$ are:

Call option

Payoff: $c_T=Max(0,S_T−X)$ <br>
Profit: $π=c_T−c_0$

Put option

Payoff: $p_T=Max(0,X−S_T)$ <br>
Profit: $π=p_T−p_0$

Options may be customized and traded in the OTC market or standardized and traded on a public exchange. Exchanges have clearinghouses to guarantee the performance of all parties and eliminate default risk. In the OTC market, the option buyer is exposed to the possibility that the seller will default if the option expires in the money.

Buyers of calls and puts have limited loss potential. If prices move against them, they cannot lose more than the amount of the premium that they have already paid. By contrast, sellers of call options are exposed to unlimited potential losses. Sellers of put options can incur substantial (but not unlimited) losses.

`European options` can only be exercised at maturity, while `American-style options` can be exercised any time after initiation. Options can be exercised by physical delivery or cash settlement.

Investors who are `bullish` can buy calls or sell puts, while `bearish` investors can buy puts or sell calls.
#### Credit Derivatives
Credit derivatives were created to hedge credit risk. The credit protection buyer gets protection from the credit protection seller. Types of credit derivatives include:
- total return swaps
- credit spread options 
- credit-linked notes
- credit default swaps

The first type of credit derivative was a total return swap. In this derivative, the credit protection seller pays a fixed or floating interest rate in return for the total return on the underlying bond or loan. If a default occurs, the credit protection seller must continue making payments.

The credit spread option is based on the bond credit spread as the underlying. The payoff is determined based on the difference between the bond’s yield spread and the strike rate. From the investor's perspective, the option is in the money if the bond's yield spread is greater than the strike rate.

With a credit-linked note, a credit protection buyer issues a note that has a principal payoff tied to a bond that he or she owns. In the event of a default on that bond, the obligations of the credit-linked note are reduced accordingly.

The most popular type of credit derivative is the `credit default swap (CDS)`. The credit protection buyer makes a series of cash payments in return for compensation from the credit protection seller if a defined loss event occurs. The structure of a CDS is effectively the same as an insurance policy.

The CDS seller is hoping the premiums collected will be greater than the payments made. Sellers should strive to diversify like insurance companies. This, unfortunately, was not the case in the financial crisis that began in 2007. The correlations among the borrowers were substantial.
#### Asset-backed Securities
An asset-backed security represents a claim on a pool of securities. A `collateralized mortgage obligation (CMO)` is a popular ABS that passes payments on underlying mortgages to investors in various tranches. CMO investors are exposed to credit risk if there are defaults on any of the underlying mortgages and prepayment risk if homeowners choose to refinance their fixed-rate mortgages.

`Collateralized debt obligations (CDOs)` are another type of ABS with bonds or loans as their underlying assets. Historically, CDOs have less prepayment risk and more credit risk than CMOs. Junior CMO tranches absorb the first credit losses, so they are priced to offer the highest expected returns. Senior CMO tranches are the most insulated from credit losses, but they offer the lowest expected returns.

Buying an ABS is equivalent to writing a put on the underlying securities. Investors will receive their promised cash flows if the underlying assets perform, but they will incur losses in the event of defaults or prepayments.
#### Hybrids
In addition to the basic categories of derivatives described in this section, hybrid instruments can be created by combining derivatives with other derivatives or securities. A corporate bond with an embedded call option (callable debt) is an example of a hybrid instrument. Swap contracts that grant one part the option to exercise (swaptions) are hybrid instruments created from two different types of derivatives.

## Derivative Underlyings
#### Equities
Both individual stocks and equity indexes are commonly used as underlying assets for various types of derivatives contracts. Index swaps are often used to increase or decrease exposure to equities as an asset class without having to trade the underlying securities. Stock options are often used as a form of incentive-based compensation for corporate executives.
#### Fixed-Income Instruments and Interest Rates
It is difficult to create derivatives on bonds because most issuers have more than one issue outstanding. This is often handled by allowing multiple issues to be delivered in order to settle contracts. Although many derivatives are based on interest rates, these are not actually assets in the same sense as equities because it is not possible to own an interest rate.
#### Currencies
The foreign exchange market is extremely large and derivatives have been very useful in helping companies and investors manage their currency risk exposures. Foreign currency positions also create exposure to interest rate risk. Forwards, futures, swaps, and options are all widely used in the currency derivatives market.
#### Commodities
Commodities include natural resources (e.g., corn, gold) that are used to sustain life or support economic activity. Many commodities (e.g., oil) have large markets and prices often react sharply to global events. While there are forward contracts, swaps, and options based on various commodities, futures contracts are the most common derivative in this sector.
#### Credit
Although credit is clearly not an underlying "asset" like securities or physical commodities, credit derivatives are popular instruments for managing exposure to credit risk.
#### Other
Other "assets" that serve as the underlying for derivatives contracts include weather, electricity, and disasters. New derivative products will emerge as long as there is sufficient interest among traders and risk managers.

### The Size of the Derivatives Market
It is difficult to estimate the exact size of the derivative market because only exchange-traded transactions are made public. OTC derivatives are not accurately tracked. Also, while the notional principal of all exchange-traded contracts can be tracked, this amount does not correspond with the monetary value of transactions or the amount at risk.

## Purposes and Benefits 
`Risk management` is the process by which an organization or individual defines the level of risk it wishes to take, measures the level of risk it is taking, and adjusts the latter to equal the former. 

Derivative markets have been used to facilitate commerce for several centuries. The first modern futures market, the Chicago Board of Trade (CBT), was founded in 1848 to help farmers achieve price certainty when selling their crops. Over time, new futures contracts have been created for other agricultural products and financial assets.

The purposes of derivative markets include the following:
#### Risk Allocation, Transfer, and Management
Derivatives can be used to get from the actual risk level to the desired risk level. Market participants can adjust their risk exposure to underlying assets without having to trade them directly, which is a more efficient method of risk management.
#### Information Discovery
Futures prices provide information on the underlying assets. The contract with the shortest maturity gives a reasonable indication of the spot price of the underlying asset. Because derivatives positions can be established with relatively small capital commitments, these markets are efficient in the sense that prices adjust quickly to reflect new information.

Prices in futures markets reveal what investors would accept today to avoid uncertainty over the term of the contract. Option prices can be used to derive the expected (or implied) volatility of underlying assets.
#### Operational Advantages
Compared to spot markets for underlying assets, derivatives markets generally have lower transaction costs and greater liquidity. Derivatives also make it much easier for investors to take short positions.
#### Market Efficiency
Lower capital requirements and transaction costs in derivatives markets make it easier to exploit mispricings, so assets trade closer to their intrinsic value.
## Criticisms and Misuses 
#### Speculation and Gambling
Using derivatives to hedge risk exposure is relatively uncontroversial. For example, there is little call for laws that would prohibit farmers from trading futures contracts to protect against the possibility of falling commodity prices.

By contrast, speculators are often perceived as trading for short-term gains without regard to long-term market stability. According to their critics, speculators are effectively engaged in legalized gambling that distorts prices in ways that adversely affect consumers and the overall economy.

However, in order for parties to hedge their risks, there must be speculators willing to accept these risks. If the party that sells a put contract to give a farmer certainty about wheat prices is a speculator, then so is the insurance company that agrees to cover that farmer's property against damages due to fires or flooding. Those opposing derivatives trading on the grounds that it is "speculation" must explain how this activity is different than other forms of financial risk-taking.
#### Destabilization and Systematic Risk
Some argue the ease of derivative trading creates the potential for catastrophic losses leading to a series of defaults that inflict significant harm on the overall economy. This concern is not unfounded, as governments have been forced to intervene to contain economic crises that were, at least in part, attributable to the use of derivatives.

Of course, there have been financial crises that pre-date the modern era of derivatives trading. But there is a sense among many that derivative trading is more sinister than other advanced sciences due to the complexity of these instruments and the potential to cause widespread risk throughout the economy. These critics believe that regulating derivatives trading would prevent (or at least limit) artificially high prices (bubbles) and subsequent corrections (crashes).

## Elementary Principles of Derivative Pricing
Basic principles of derivative pricing and the concept of arbitrage. 

Like stocks, bonds, and other assets, derivatives can be valued by discounting their expected future cash flows. However, the discount rate used to value a derivative should not be the same as the one used to value its underlying asset because the effect of this discount rate is already reflected in the price of the underlying. Rather, derivatives should be valued based on the premise that they can be used to create positions that are completely hedged. A position with zero risks should earn the risk-free rate.
### Storage
Some commodities (e.g., gold) can effectively be stored indefinitely, while others (e.g., bananas) can only be stored for a short period before they spoil. Certain commodities (e.g., electricity) cannot be stored at all.

Storage costs can be significant for some assets, but they are effectively zero for financial assets. Investors who sell financial assets forward should expect to earn the `cost of carry`, which is the risk-free rate less the dividends received during the term of the contract.
### Arbitrage
Derivative prices are set to avoid arbitrage. The simplest case of arbitrage is a single asset selling for different prices simultaneously, which is a violation of the law of one price.

In theory, assets that offer the same benefits and have the same risk should sell for the same price. If the law of one price is violated, investors will act quickly to exploit arbitrage opportunities and bring the prices back in line. However, assets may sell at different prices in different locations if the transportation costs exceed the potential arbitrage profit.

If markets are efficient, there should be relatively few arbitrage opportunities. However, because derivative valuations are heavily influenced by the underlying asset's volatility, which can only be estimated, different investors may perceive discrepancies between the prices of derivatives relative to those of their underlying assets.

Because arbitrage is a relative valuation methodology, it cannot be used to determine whether an asset's price is objectively correct. Rather, arbitrage can indicate whether an asset's price is fair relative to the price of another asset or derivative.
