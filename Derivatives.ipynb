# Derivative Markets and Instruments
`Derivatives`—financial instruments whose prices are derived from the value of some underlying asset (referred to as the `underlying`) —have become increasingly important for managing financial risk, exploiting investment opportunities, and creating synthetic asset class exposure. As in other security markets, arbitrage and market efficiency play a critical role in establishing prices for these securities. They have a defined and limited life.

## Introduction
Perhaps the distinction that best characterizes derivatives is that they usually transform the performance of the underlying asset before paying it out in the derivatives transaction.

Derivatives are similar to insurance in that both allow for the transfer of risk from one party to another. The buyer of a derivative contract has a `long position` and the seller has a `short position`.

*(Note: Investors maintain “long” security positions in the expectation that the stock will rise in value in the future. The opposite of a “long” position is a “short” position. A "short" position is generally the sale of a stock you do not own. Investors who sell short believe the price of the stock will decrease in value.)*

The so-called “underlying asset” is often simply referred to as the underlying, whose value is the **source of risk**. Common derivatives underlyings are:
- equities
- fixed-income securities 
- currencies
- commodities

Other derivatives underlyings include:
- interest rates
- credit
- energy
- weather
- other derivatives

Thus, like insurance, derivatives pay off on the basis of a source of risk, which is often, but not always, the value of an underlying asset. And like insurance, derivatives have a definite life span and expire on a specified date.
#### Classes (or Forms)
There are two general classes of derivatives:
1. Some provide the ability to lock in a price at which one might buy or sell the underlying. Because they force the two parties to transact in the future at a previously agreed-on price, these instruments are called `forward commitments`: 
  - forward contracts
  - futures contracts
  - swaps
2. `Contingent claims` provides the right but not the obligation to buy or sell the underlying at a pre-determined price (because the choice of buying or selling versus doing nothing depends on a particular random outcome, these derivatives are called contingent claims):
  - options

#### Purposes
Derivatives can be used to create strategies that would not exist in their absence. For example, derivatives make it easier to go short, thereby benefiting from a decline in the value of the underlying.

Leverage can also be easily created because investments are small relative to the value of the underlying. As such, small movements in the underlying can lead to fairly large movements in the amount of money made or lost on the derivative.

Derivatives usually have lower transaction costs than comparable spot market transactions and more liquidity compared to their underlying asset.  

They are critical to the risk management process because they can be used to adjust, reduce, or eliminate exposure to certain risks easily and quickly. *(For example, a shareholder of a company can reduce or even completely eliminate the market exposure by trading a derivative on the equity. Holders of fixed-income securities can use derivatives to reduce or completely eliminate interest rate risk, allowing them to focus on the credit risk. Alternatively, holders of fixed-income securities can reduce or eliminate the credit risk, focusing more on the interest rate risk.)*  


## Derivative Market Structure
Certain derivatives trade on `public exchanges` while others are traded in the `over-the-counter (OTC)` market.
### Exchange-Traded Derivatives Markets
Exchange-traded derivatives are standardized contracts, which means variations in terms such as expiration and underlying securities are limited. Standardized terms increase market liquidity because all investors are trading the same contract. `Market makers` and `speculators` are active participants on derivatives exchanges.

Clearing and settlement operations are smooth and come with a `credit guarantee`. The clearinghouse is able to guarantee the losing party's performance by requiring margin bonds of the participants. Exchange markets are also transparent. This information is useful to many but also results in a loss of privacy.
### Over-The-Counter Derivatives Markets
The OTC derivatives market operates through an informal network of dealers (usually banks) that buy and sell contracts at their discretion. OTC derivatives contracts are not standardized, which creates more risk for the dealer because it is difficult to find identical contracts to offset risk. Instead, dealers manage their risk exposures by taking offsetting positions in similar contracts.

OTC contracts are not universally illiquid. Some can have higher trading volumes than certain exchange-traded derivatives contracts. The OTC derivatives market has traditionally been lightly regulated, which gives participants more privacy and flexibility.

## Types of Derivatives
### Forward Commitments
Forward commitments can be exchange-traded contracts (e.g., futures) or over-the-counter (OTC) contracts (e.g., forwards and swaps).
#### Forward Cotracts
A forward contract is an agreement in advance to make a trade. One party agrees to pay the `forward price` for an underlying asset and the other party agrees to sell at that price on a specific date. *(The buyer is called the long party and the seller is called the short party.)*

Contract terms (e.g., where and when delivery will be made, the quantity of the underlying asset to be delivered) can be **customized** to meet the specific needs of each party. Forward contracts are regularly written on a wide variety of underlying assets.

Forward contracts have **zero value at inception**, so no money is exchanged at that time. They become more or less valuable to the respective parties as the price of the underlying asset fluctuates during the term of the contract. The value of a forward contract depends on the type of position:

To illustrate the payoff of a forward contract, start with the assumption that we are at time $t = 0$ and that the forward contract expires at a later date, time $t = T$. The `spot price` of the underlying asset at time 0 is $S_0$ and at time $T$ is $S_T$. The notation $F_0(T)$ denotes that this value is established at time $0$ and applies to a contract expiring at time $T$. $F_0(T)$ is the forward price. Roll forward to time $T$, when the price of the underlying is $S_T$. The long is obligated to pay $F_0(T)$, for which he receives an asset worth $S_T$. 

Long: $S_T−F_0(T)$<br>
Short: $−[S_T−F_0(T)]=F_0(T)−S_T$

As shown in the payoff formulas, the long party gains if the price of the underlying asset is above the forward price at expiry. Forward contracts are a zero-sum game because the long party's gain is equal to the short party's loss (or vice versa if the price of the underlying finishes below the forward price). The gain from owning the underlying would be $S_T ‒ S_0$, whereas the gain from owning the forward contract would be $S_T ‒ F_0(T)$. 

Each party faces the risk the other party will default, although only the "winning" party is exposed to default risk at any given time during the life of the contract. This usually limits participation to creditworthy entities. The forward market is private and largely unregulated.

Forward contracts can be used to hedge various risks. For example, manufacturers can use forward contracts to lock in the price of their raw materials. Many forward contracts require the seller to physically deliver the underlying asset to the long party. `Non-deliverable forwards (NDFs)` do not require physical delivery. Instead, the losing party makes a cash payment to the winning party for the amount of the difference between the forward price and the spot price of the underlying asset at expiry.
#### Futures
Futures contracts are specialized versions of forward contracts that have been standardized and that trade on a futures exchange. A futures contract is similar to a forward contract in that there is an agreement to purchase an underlying asset for a specific price on a specified date. Differences between forwards and futures include:
- Standardization: Futures contracts are `standardized`, while forward contracts are customized terms.
- Futures contracts trade on a `public exchange` with a `clearinghouse` that guarantees the performance of all traders. Forward contracts trade OTC with no performance guarantees, so traders must do their own assessment of counterparty credit risk.
- `Daily settlement`: Futures contracts are `marked to market`, which allows traders to realize gains or losses at the end of each day. By contrast, gains and losses for forward contracts are not realized until expiration. (Futures contracts are more liquid because it is much easier to take offsetting positions when trading standardized contracts. By contrast, forward contracts are usually held to maturity.)

The account is specifically referred to as a `margin account`. A `margin call` is made if an account balance drops below the `maintenance margin`. After a margin call, the party must deposit sufficient funds to bring the account balance back to the initial margin level. Note that it is not sufficient to bring the balance back to the maintenance margin.

    Note: (Margin account is used for an amount of money that a trader deposits in a margin account. The term is derived from the stock market practice in which an investor borrows a portion of the money required to purchase a certain amount of stock. In futures markets, there is no borrowing so the margin is more of a down payment or performance bond.)

Futures exchanges minimize their losses by requiring an `initial margin deposit` (typically less than 10% of the futures price) from both parties upon initiation. It serves as a performance bond to provide some assurance of the parties' ability to cover possible future losses. At the end of each trading day, gains are credited to one party's margin account based on the change in the futures price. An equivalent amount is debited from the balance of the other party's account.

**Price limits:** Some futures contracts have price limits that restrict the trading range/band. If prices reach these limits (limit up), trading will temporarily stop until a trade takes place above the `limit down` price or below the `limit up` price.

**Settlement:** Like forward contracts, futures may be settled in cash or with physical delivery of the underlying asset. Regardless of the settlement terms, **the futures price will `converge` to the spot price at expiration**.

    In the futures markets, the exchange specifies whether physical delivery or cash settlement applies. In physical delivery contracts, the short is required to deliver the underlying asset at a designated location and the long is required to pay for it. Delivery replaces the mark-to-market process on the final day. It also ensures an important principle that you will use later: The futures price converges to the spot price at expiration. Because the short delivers the actual asset and the long pays the current spot price for it, the futures price at expiration has to be the spot price at that time. Alternatively, a futures contract initiated right at the instant of expiration is effectively a spot transaction and, therefore, the futures price at expiration must equal the spot price. Following this logic, in cash settlement contracts, there is a final mark to market, with the futures price formally set to the spot price, thereby ensuring automatic convergence.

At any given time, the number of outstanding contracts is called the `open interest`. Each contract counted in the open interest has a long and a corresponding short. 

**Payoff:** The total payoff for a futures contract is the same as for a forward contract (see detailed deduction in textbook, which is straightforward), but the timing is different. Forward contracts realize the full amount at maturity, while futures contracts make and receive cash flows daily. (related: what's called an "offset".)

Futures prices of nearby expiring contracts are often used as proxies for spot prices, particularly in decentralized spot markets, such as gold, which trades in spot markets all over the world. (One of the ways how derivatives affect its underlying's values?)
#### Swaps
The concept of a swap is that two parties exchange (swap) a series of cash flows. One set of cash flows is variable or floating and will be determined by the movement of an underlying asset or rate. The other set of cash flows can be variable and determined by a different underlying asset or rate, or it can be fixed. 

A swap is a series of forward contracts, meaning that cash flows are exchanged on more than one date. Like forward contracts, swaps are private OTC transactions and each party is exposed to the risk that the other party will default. (Combinations of futures contracts expiring at different dates are often compared to swaps.)

They are often used by parties seeking to convert their `fixed-rate debt` into `floating-rate obligations` (or vice versa). This is called a `fixed-for-floating interest rate swap`, or a `plain vanilla swap`.

As with futures and forwards, no money changes hands at the start; thus, the value of a swap when initiated must be zero. The fixed rate on the swap is determined by a process that forces the value to zero, a procedure that will be covered later in the curriculum. As market conditions change, the value of a swap will deviate from zero, being positive to one party and negative to the other.

A swap has no net value at inception, so no money is exchanged when the contract is signed. Periodic payments are made based on the contract's `notional principal` and the rates that each party has agreed to pay. On each payment date, a net payment is made by the party owing more. (The swap has no principal, but it does have a balance of sorts, called the notional principal, which ordinarily matches the loan balance. A loan with only one principal payment, the final one, will be matched with a swap with a fixed notional principal. An amortizing loan, which has a declining principal balance, will be matched with a swap with a pre-specified declining notional principal that matches the loan balance.)

### Contingent Claims
A contingent claim is a derivative in which the outcome or payoff is dependent on the outcome or payoff of an underlying asset. For contingent claims, the payoff occurs if a specific event happens. Examples include options, credit derivatives, and asset-backed securities.
#### Options
An option gives one party the right, but not the obligation, to buy or sell an underlying asset. The terms of an option (e.g., the underlying asset, exercise price, date, quantity) are specified in advance.

A call option grants the right to buy the underlying asset and a put option grants the right to sell. The party that buys an option (either a call or a put) is called the long party, while the option seller is called the short party. For example, an investor who sells a put is said to have a short put position.

*Unlike forward commitments, a **cash payment** is made when the contract is signed. The right to buy or sell will only be exercised if doing so would be profitable for the party holding the option. If not, the option contract will be allowed to expire unused. The ability to profit from good outcomes and walk away from bad outcomes is valuable, so the option buyer is required to make an upfront payment (called a premium) to the option seller.*

The value of an option after initiation depends on the price of the underlying asset relative to the `exercise price`. For example, if a stock is trading for $40, a call option with an exercise price of \\$35 is described as `being in the money` because the call buyer has the right to buy the stock for less than its market value. The same option would be described as `being at the money` if the stock was trading at the \$35 exercise price or `out of the money` if it was trading below that level.

    The fixed price at which the underlying asset can be purchased is called the exercise price (also called the “strike price,” the “strike,” or the “striking price”). 

If $X$ is the exercise price and $c_0$ and $p_0$ are the premiums for call and put options, respectively, then the payoff and profit to the call and put buyers at maturity time $T$ are:

Call option

Payoff: $c_T=Max(0,S_T−X)$ <br>
Profit: $π=c_T−c_0$

Put option

Payoff: $p_T=Max(0,X−S_T)$ <br>
Profit: $π=p_T−p_0$

Options may be customized and traded in the OTC market or standardized and traded on a public exchange. Exchanges have clearinghouses to guarantee the performance of all parties and eliminate default risk. In the OTC market, the option buyer is exposed to the possibility that the seller will default if the option expires in the money.

Buyers of calls and puts have limited loss potential. If prices move against them, they cannot lose more than the amount of the premium that they have already paid. By contrast, sellers of call options are exposed to unlimited potential losses. Sellers of put options can incur substantial (but not unlimited) losses.

`European options` can only be exercised at maturity, while `American-style options` can be exercised any time after initiation. Options can be exercised by physical delivery or cash settlement.

Investors who are `bullish` can buy calls or sell puts, while `bearish` investors can buy puts or sell calls.
#### Credit Derivatives
Credit derivatives were created to hedge credit risk. The credit protection buyer gets protection from the credit protection seller. Types of credit derivatives include:
- total return swaps
- credit spread options 
- credit-linked notes
- credit default swaps

The first type of credit derivative was a total return swap. In this derivative, the credit protection seller pays a fixed or floating interest rate in return for the total return on the underlying bond or loan. If a default occurs, the credit protection seller must continue making payments.

The credit spread option is based on the bond credit spread as the underlying. The payoff is determined based on the difference between the bond’s yield spread and the strike rate. From the investor's perspective, the option is in the money if the bond's yield spread is greater than the strike rate.

With a credit-linked note, a credit protection buyer issues a note that has a principal payoff tied to a bond that he or she owns. In the event of a default on that bond, the obligations of the credit-linked note are reduced accordingly.

The most popular type of credit derivative is the `credit default swap (CDS)`. The credit protection buyer makes a series of cash payments in return for compensation from the credit protection seller if a defined loss event occurs. The structure of a CDS is effectively the same as an insurance policy.

The CDS seller is hoping the premiums collected will be greater than the payments made. Sellers should strive to diversify like insurance companies. This, unfortunately, was not the case in the financial crisis that began in 2007. The correlations among the borrowers were substantial.
#### Asset-backed Securities
An asset-backed security represents a claim on a pool of securities. A `collateralized mortgage obligation (CMO)` is a popular ABS that passes payments on underlying mortgages to investors in various tranches. CMO investors are exposed to credit risk if there are defaults on any of the underlying mortgages and prepayment risk if homeowners choose to refinance their fixed-rate mortgages.

`Collateralized debt obligations (CDOs)` are another type of ABS with bonds or loans as their underlying assets. Historically, CDOs have less prepayment risk and more credit risk than CMOs. Junior CMO tranches absorb the first credit losses, so they are priced to offer the highest expected returns. Senior CMO tranches are the most insulated from credit losses, but they offer the lowest expected returns.

Buying an ABS is equivalent to writing a put on the underlying securities. Investors will receive their promised cash flows if the underlying assets perform, but they will incur losses in the event of defaults or prepayments.
#### Hybrids
In addition to the basic categories of derivatives described in this section, hybrid instruments can be created by combining derivatives with other derivatives or securities. A corporate bond with an embedded call option (callable debt) is an example of a hybrid instrument. Swap contracts that grant one part the option to exercise (swaptions) are hybrid instruments created from two different types of derivatives.

## Derivative Underlyings
#### Equities
Both individual stocks and equity indexes are commonly used as underlying assets for various types of derivatives contracts. Index swaps are often used to increase or decrease exposure to equities as an asset class without having to trade the underlying securities. Stock options are often used as a form of incentive-based compensation for corporate executives.

It is difficult to create derivatives on bonds because most issuers have more than one issue outstanding. #### Currencies
The foreign exchange market is extremely large and derivatives have been very useful in helping companies and investors manage their currency risk exposures. Foreign currency positions also create exposure to interest rate risk. Forwards, futures, swaps, and options are all widely used in the currency derivatives market.
#### Commodities
Commodities include natural resources (e.g., corn, gold) that are used to sustain life or support economic activity. Many commodities (e.g., oil) have large markets and prices often react sharply to global events. While there are forward contracts, swaps, and options based on various commodities, futures contracts are the most common derivative in this sector.
#### Credit
Although credit is clearly not an underlying "asset" like securities or physical commodities, credit derivatives are popular instruments for managing exposure to credit risk.
#### Other
Other "assets" that serve as the underlying for derivatives contracts include weather, electricity, and disasters. New derivative products will emerge as long as there is sufficient interest among traders and risk managers.

## Purposes and Benefits 
`Risk management` is the process by which an organization or individual defines the level of risk it wishes to take, measures the level of risk it is taking, and adjusts the latter to equal the former. 

Derivative markets have been used to facilitate commerce for several centuries. The first modern futures market, the Chicago Board of Trade (CBT), was founded in 1848 to help farmers achieve price certainty when selling their crops. Over time, new futures contracts have been created for other agricultural products and financial assets.

The purposes of derivative markets include the following:
#### Risk Allocation, Transfer, and Management
Derivatives can be used to get from the actual risk level to the desired risk level. Market participants can adjust their risk exposure to underlying assets without having to trade them directly, which is a more efficient method of risk management.
#### Information Discovery
Futures prices provide information on the underlying assets. The contract with the shortest maturity gives a reasonable indication of the spot price of the underlying asset. Because derivatives positions can be established with relatively small capital commitments, these markets are efficient in the sense that prices adjust quickly to reflect new information.

Prices in futures markets reveal what investors would accept today to avoid uncertainty over the term of the contract. Option prices can be used to derive the expected (or implied) volatility of underlying assets.
#### Operational Advantages
Compared to spot markets for underlying assets, derivatives markets generally have lower transaction costs and greater liquidity. Derivatives also make it much easier for investors to take short positions.
#### Market Efficiency
Lower capital requirements and transaction costs in derivatives markets make it easier to exploit mispricings, so assets trade closer to their intrinsic value.
## Criticisms and Misuses 
#### Speculation and Gambling
Using derivatives to hedge risk exposure is relatively uncontroversial. For example, there is little call for laws that would prohibit farmers from trading futures contracts to protect against the possibility of falling commodity prices.

By contrast, speculators are often perceived as trading for short-term gains without regard to long-term market stability. According to their critics, speculators are effectively engaged in legalized gambling that distorts prices in ways that adversely affect consumers and the overall economy.

However, in order for parties to hedge their risks, there must be speculators willing to accept these risks. If the party that sells a put contract to give a farmer certainty about wheat prices is a speculator, then so is the insurance company that agrees to cover that farmer's property against damages due to fires or flooding. Those opposing derivatives trading on the grounds that it is "speculation" must explain how this activity is different than other forms of financial risk-taking.
#### Destabilization and Systematic Risk
Some argue the ease of derivative trading creates the potential for catastrophic losses leading to a series of defaults that inflict significant harm on the overall economy. This concern is not unfounded, as governments have been forced to intervene to contain economic crises that were, at least in part, attributable to the use of derivatives.

Of course, there have been financial crises that pre-date the modern era of derivatives trading. But there is a sense among many that derivative trading is more sinister than other advanced sciences due to the complexity of these instruments and the potential to cause widespread risk throughout the economy. These critics believe that regulating derivatives trading would prevent (or at least limit) artificially high prices (bubbles) and subsequent corrections (crashes).

## Elementary Principles of Derivative Pricing
Basic principles of derivative pricing and the concept of arbitrage. 

Like stocks, bonds, and other assets, derivatives can be valued by discounting their expected future cash flows. However, the discount rate used to value a derivative should not be the same as the one used to value its underlying asset because the effect of this discount rate is already reflected in the price of the underlying. Rather, derivatives should be valued based on the premise that they can be used to create positions that are completely hedged. A position with zero risks should earn the risk-free rate.
### Storage
Some commodities (e.g., gold) can effectively be stored indefinitely, while others (e.g., bananas) can only be stored for a short period before they spoil. Certain commodities (e.g., electricity) cannot be stored at all.

Storage costs can be significant for some assets, but they are effectively zero for financial assets. Investors who sell financial assets forward should expect to earn the `cost of carry`, which is the risk-free rate less the dividends received during the term of the contract.
### Arbitrage
Derivative prices are set to avoid arbitrage. The simplest case of arbitrage is a single asset selling for different prices simultaneously, which is a violation of the law of one price.

In theory, assets that offer the same benefits and have the same risk should sell for the same price. If the law of one price is violated, investors will act quickly to exploit arbitrage opportunities and bring the prices back in line. However, assets may sell at different prices in different locations if the transportation costs exceed the potential arbitrage profit.

If markets are efficient, there should be relatively few arbitrage opportunities. However, because derivative valuations are heavily influenced by the underlying asset's volatility, which can only be estimated, different investors may perceive discrepancies between the prices of derivatives relative to those of their underlying assets.

Because arbitrage is a relative valuation methodology, it cannot be used to determine whether an asset's price is objectively correct. Rather, arbitrage can indicate whether an asset's price is fair relative to the price of another asset or derivative.

# Basics of Derivative Pricing and Valuation

## Fundamental Concepts of Derivative Pricing
### Pricing the Underlying
Derivatives are priced based on an underlying asset. The four main types of underlying assets are 
- equities, 
- fixed-income securities/interest rates, 
- currencies, and 
- commodities. 
<br>Although interest rates are not technically assets, they are used as the underlying for many derivatives.

We use the symbol $k$ to denote this currently unknown discount rate, which is often referred to as the required rate of return and sometimes the expected rate of return or just the expected return. At a minimum, that rate will include the risk-free rate of interest, which we denote as $r$. This rate represents the opportunity cost, or so-called time value of money, and reflects the price of giving up your money today in return for receiving more money later.

Assets are often valued as the present value of their expected future cash flows. Of course, this requires a forecast of the cash flows and assumptions about the discount rate, which can be decomposed into the risk-free rate ($r$) and a premium ($λ$) that investors require as compensation for accepting various risks.

The premium is influenced by the riskiness of the asset's expected cash flows and the investor’s degree of risk aversion. Investors can be classified as `risk-averse`, `risk-neutral`, or `risk-seeking`.  
- Risk-neutral investors
  - are ambivalent to risk, thus expect to earn the risk-free rate on every investment, regardless of the risk.
  - Risk-neutral investors neither give nor receive a risk premium because they have no feelings about risk.
- Risk seekers 
  - actually enjoy risk, so they would accept a lower return on a risky investment for the privilege of taking on risk.
  - Risk-seeking investors give away a risk premium because they enjoy taking risk.
- Risk-averse investors 
  - will demand a greater expected return on investments that have more risk. So $λ$ will be $0$ for risk-free investments and positive for all other investments. The traditional financial theory is based on the assumption that investors are risk-averse.
  - Risk-averse investors expect a risk premium to compensate for the risk. $E(r)=R_f+\beta\times MRP$ (the expected return for a stock is the risk free rate plus beta times market risk premium)

Assuming a financial asset has no interim cash flows, the current price (i.e., spot price), $S_0$, is equal to the expected future price, $E(S_T)$, discounted back to time $0$.

$$S_0=\frac{E(S_T)}{(1+r+λ)^T}$$

**Benefits of Holding an Asset:** Many times investors of financial assets receive `dividends or interest payments`. Some commodities provide a `convenience yield` for the holder. This nonmonetary benefit could be due to the pleasure of holding something (like gold) or the security of having it if a supply shortage occurs. However, holding an asset also imposes storage costs and opportunity costs that can partially or completely offset the convenience yield.

Assuming $θ$ represents the present value of `costs of holding an asset` (e.g., storage) and $γ$ represents the present value of benefits, the asset price formula can be revised to the following:

$$S_0=\frac{E(S_T)}{(1+r+λ)^T}−θ+γ$$

### The Principle of Arbitrage
Arbitrage is a type of transaction undertaken when two assets or portfolios produce identical results but sell for different prices. Arbitrage opportunities exist if a trader can buy an asset at one price and simultaneously sell an identical asset at a greater price. Investors who are able to do this will earn a profit without taking any risk. 

*(Note: Because the two assets or portfolios produce identical results, a long position in one and a short position in the other means that at the end of the holding period, the payoffs offset. Hence, no money is gained or lost at the end of the holding period, so there is no risk)*. 

The net effect is that the arbitrageur receives money **at the start** and never has to pay out any money later. Such a situation amounts to free money.

Such opportunities are rare and last for only short periods of time before arbitrage trading corrects the mispricing.
#### Arbitrage and Derivatives
All of the uncertainty in a derivative comes from the uncertainty in the underlying. As a result, the price of the derivative is tied to the price of the underlying. That being the case, the derivative can be used to **hedge the underlying, or vice versa**.

When a long position in the underlying is combined with a short position in the derivative to produce a perfect `hedge`, all of the risk is eliminated and the position should earn the risk-free rate. If not, arbitrageurs begin to trade. If the position generates a return in excess of the risk-free rate, the arbitrageurs see an opportunity because the hedged position of the underlying (long asset and short derivative) earns more than the risk-free rate and a risk-free loan undertaken as a borrower incurs a cost equal to the risk-free rate. Therefore, implementing the hedged position and borrowing at the risk-free rate earns a return in excess of the risk-free rate, incurs a cost of the risk-free rate, and has no risk. As a result, an investor can earn excess return at no risk without committing any capital. 

Arbitrageurs will execute this transaction in large volumes, continuing to exploit the pricing discrepancy until market forces push prices back in line such that both risk-free transactions earn the risk-free rate.
#### Arbitrage and Replication
Because a derivative's value is directly linked to the price of its underlying asset, arbitrage opportunities are possible if the two are not priced consistently. Investors can take advantage of mispricings by creating synthetic positions that combine a derivative, its underlying asset, and a risk-free security.

These `replication strategies` are summarized below:
$$Long\ Asset + Short\ Derivative = Long\ Riskfree\ Asset\ (Lending)$$
$$Long\ Asset+ Short\ Riskfree\ Asset\ (Borrowing)=Long\ Derivative$$
$$Short\ Derivative+Short\ Riskfree\ Asset\ (Borrowing)=Short\ Asset$$

(in a forward contract, a long position means you are obligated to buy the underlying)

Asset + Derivative = Risk-free asset<br>
Asset−Risk-free asset=−Derivative<br> 
Derivative−Risk-free asset=−Asset

   
    It may be more profitable to hedge a portfolio with a derivative to produce a risk-free rate than to invest in the risk-free asset. In addition, we might find that replication can have lower transaction costs. For example, a derivative on a stock index combined with the risk-free asset [Long derivative (Stock index futures) + Long risk-free asset (Lending) = Long asset (Stock index)] can potentially replicate an index fund at lower transaction costs than buying all the securities in the index. Replication is the essence of arbitrage.
    
It may be impossible to exploit apparent arbitrage opportunities if the transaction costs are greater than the magnitude of the mispricing. Even if arbitrage opportunities persist after accounting for transaction costs, it may be difficult to raise the capital needed to exploit them by borrowing at the risk-free rate. In general, arbitrage can be exploited when there are:
- Low transaction costs
- High liquidity
- A large price differential

Overall, finding and exploiting arbitrage opportunities is an expensive and time-consuming endeavor.

## The Concept of Pricing vs. Valuation
**The value is the amount of wealth represented by owning the forward contract. The price is one of the terms the parties agreed on when they created the contract.**

`Fundamental valuation` approaches can be used to determine appropriate prices for stocks. If the fundamental value differs significantly from the current price, traders may be enticed to buy or sell the security.

This same approach can be used in the derivatives market, but the terminology is a bit different. For example, forwards, futures, and swaps do not require cash outlays at inception. The price for these contracts refers to the price at which the underlying will be traded at a later date. It is not equivalent to the value of the contract at the start.
### Risk-neutral Pricing
**`Risk-neutral pricing` (a.k.a. `arbitrage-free pricing`, or `principle of no arbitrage`)**: As noted, it is reasonable to assume that investors are risk-averse. However, derivative prices should not be affected by an investor’s degree of risk aversion. **Derivatives are priced by discounting payoffs at the `risk-free rate`** because the asset price has already built in the risk discount. 

Risk-neutral pricing uses the fact that arbitrage opportunities guarantee that a risk-free portfolio consisting of the underlying and the derivative must earn the risk-free rate. There is only one derivative price that meets that condition. Any mispricing of the derivative will lead to arbitrage transactions that drive the derivative price back to where it should be, the price that eliminates arbitrage opportunities.

## Pricing and Valuation: Forward Commitments
### Forward Contracts

#### Pricing and Valuation at Expiration
Forward commitments are created at time $0$,

Expiration at time $T$,

The value of the underlying today is $S_0$,

At expiration the underlying value is $S_T$, which is not known at the initiation of the contract,

The value at expiration is $V_T(T)$,

The price at which to transact the underlying security $S$ is agreed upon at inception and is denoted as $F_0(T)$.

$$V_T(T)=S_T−F_0(T)$$
<div style="text-align: right"> Equation 1 </div>
I.e., the `value of a forward contract` at expiration is the spot price of the underlying minus the forward price agreed to in the contract. This is the value of the `long position`, which is the party buying the underlying security at the expiration date. *(Note: In the financial world, we generally define value as the value to the long position.)*

The value to the short party is the opposite, $−[S_T−F_0(T)]$.
#### Pricing and Valuation at Initiation
No cash is exchanged at inception for a forward contract. The forward price that the parties agree to at the initiation date of the contract is a special price that results in the contract having zero value and prohibiting arbitrage. Because neither the long nor the short pays anything to the other at the initiation date of a forward contract, the value of a forward contract when initiated is zero.

$$V_0(T)=0$$
<div style="text-align: right"> Equation 2 </div>

A forecast of the price of this asset at our horizon date of time $T$. That forecast was previously referred to as the expected spot price at expiration, $E(S_T)$.

If a trader owns an asset now worth $S_0$ and also enters a forward contract to sell the asset at time $T$ for $F_0(T)$, the trader effectively has a risk-free portfolio. Because there is no price risk for the transaction that will take place, the forward price must reflect the current spot price grown at the risk-free rate:

$$\frac{F_0(T)}{S_0}=(1+r)^T$$
<div style="text-align: right"> Equation 3 </div>

$$F_0(T)=S_0(1+r)^T$$
<div style="text-align:right"> Equation 4 </div>
In words, the forward price is the spot price compounded at the risk-free rate over the life of the contract.

Notice how the `risk premium` on the asset does not directly appear in the pricing relationship. It does appear implicitly, because it determines the spot price paid to buy the asset. Knowing the spot price, however, eliminates the necessity of determining the risk premium. The derivatives market can simply let the spot market derive the risk premium.

The formula for the forward price can easily be adjusted to account for benefits ($γ$) and costs ($θ$) of owning a security. All else equal, higher carrying costs or lower benefits will result in a higher forward price:
$$(1+r)^T=\frac{F_0(T)+(γ−θ)(1+r)^T}{S_0}$$

$$F_0(T)=(S_0−γ+θ)(1+r)^T$$
<div style="text-align: right"> Equation 5 </div>
or
$$F_0(T)=S_0(1+r)^T−(γ−θ)(1+r)^T$$
In other words,The forward price of an asset with benefits and/or costs is the spot price compounded at the risk-free rate over the life of the contract minus the future value of those benefits and costs.

If the net cost of carry (benefits less costs) is positive, the forward price is lower than if the net cost of carry was zero.

    Going long the forward contract enables the investor to avoid having to pay the price of the asset, S0, so she would collect interest on the money. Thus, the forward strategy would have a value of S0, reflecting the investment of that much cash invested in risk-free bonds, plus the value of the forward contract. The spot strategy would have a value of S0, reflecting the investment in the asset. These two strategies must have equal values. Hence, the value of the forward contract must be zero.
#### Pricing and Valuation during the Life of the Contract
The value of the forward contract at time $t$, which can be any time between $0$ and $T$, can be calculated as the current spot price $S_t$ of the underlying security less the present value of the forward price:

$$V_t(T)=S_t−\frac{F_0(T)}{(1+r)^{(T−t)}}$$
<div style="text-align:right"> Equation 6 </div>
Where $V_t(T)$ as the value of the forward contract at time $t$.

This equation can be amended to account for the costs and benefits of carrying the underlying asset:

$$V_t(T)=S_t−F_0(T)(1+r)^{−(T−t)}−(γ−θ)(1+r)^t$$
<div style="text-align:right"> Equation 7 </div>
Equation 7 also holds at the initiation date and at expiration. 

    The forward contract forgoes the benefits and avoids the costs of holding the asset. Consequently, we adjust the value downward to reflect the forgone benefits and upward to reflect the avoided costs. Remember that the costs (θ) and benefits (γ) are expressed on a present value basis as of time 0. We need their value at time t. We could compound them from 0 to T and then discount them back to t by the period T – t, but a shorter route is to simply compound them from 0 to t.
#### Forward Contracts on Interest Rates
Forward contracts on interest rates are called `forward rate agreements (FRA)`. Forward rate agreements are forward contracts that conceptually allow lenders to **lock in a fixed payment on a future investment** by receiving a known payment and making an unknown payment that offsets the unknown future interest payment. FRAs are unlike most derivatives in the sense that their underlying "asset" is an interest rate (e.g., Libor) rather than a tangible or financial asset. 

These agreements are widely used to manage interest rate exposure. For example, a company that intends to take out a loan in 3 months can use an FRA to lock in its cost of borrowing and protect against an increase in interest rates.

It is possible to create a `synthetic FRA` from a combination of time deposits. For example, an FRA for a 90-day loan one month from today can be replicated by combining a long position in a 120-day time deposit with a short position in a 30-day time deposit. (Going long over the 120-day period and short over the 30-day period leaves an investor with no exposure over the 30-day period and then converts to a position that starts 30 days from now and matures 90 days later. This synthetic position corresponds to a 30-day FRA on 90-day Libor.)

Even though FRAs do not involve an underlying asset, they can still be combined with an underlying asset to produce a hedged position, thereby leading to fairly straightforward pricing and valuation equations.

### Futures Contracts
Conceptually, forwards and futures contracts are very similar. In terms of cash flows, the key difference is that futures contracts are `marked to market` *(meaning the accumulated gains and losses from the previous day’s trading session are deducted from the accounts of those holding losing positions and transferred to the accounts of those holding winning positions)* on a daily basis, whereas forwards are not settled until maturity.

By requiring both parties to deposit funds in a margin account, futures exchanges guarantee their performance. Margin account balances are adjusted at the end of each trading day to reflect changes in the contract price.

**The value of a futures contract is the accumulated gain or loss on a futures contract since its previous day’s settlement. When that value is paid out in the daily settlement, the futures price is effectively reset to the settlement price and the value goes to zero.** The different patterns of cash flows for forwards and futures can lead to differences in the pricing of forwards versus futures. But there are some conditions under which the pricing is the same. It turns out that if interest rates were constant, forwards and futures would have the same prices. The differential will vary with the volatility of interest rates. 

Futures and forwards are priced differently depending on the correlation between the futures prices and interest rates:
- if futures prices and interest rates are uncorrelated, forwards and futures prices will be the same. 
- If futures prices are positively correlated with interest rates, futures contracts are more desirable to holders of long positions than are forwards. The reason is because rising prices lead to futures profits that are reinvested in periods of rising interest rates, and falling prices leads to losses that occur in periods of falling interest rates. It is far better to receive cash flows in the interim than all at expiration under such conditions. This condition makes futures more attractive than forwards, and therefore their prices will be higher than forward prices. 
- A negative correlation between futures prices and interest rates leads to the opposite interpretation, with forwards being more desirable than futures to the long position. The more desirable contract will tend to have the higher price.

The practical realities, however, are that the derivatives industry makes virtually no distinction between futures and forward prices.13 Thus, we will make no distinction between futures and forward pricing, except possibly in noting some subtle issues that may arise from time to time.

### Swap Contracts
A swap is really just a series of forward contracts. Normally, it involves a fixed payment exchanged for a floating payment. Like forward contracts, **the initial value of a swap is usually zero.** This is done by determining the fixed payment that will make the swap have a zero initial value.

Swaps can be valued by using arbitrage principles. For example, a pay-fixed, receive-floating swap can be replicated by issuing fixed-rate debt and purchasing a floating-rate bond. Since these two approaches have identical results, they should have the same value. The value of a swap after initiation can be determined from the difference between the values of the floating rate and fixed-rate bonds.

$FS_0(n, T)$ denotes the fixed payment established at time $0$ for a swap consisting of $n$ payments with the last payment at time $T$.

We denote the time to each payment as $t_1$, $t_2$, …, $t_n$, where $t_n = T$.

On each of these payment dates, the owner of the swap makes a payment of $FS_0(n, T)$ and receives a payment based on the value of the underlying at the time of each respective payment, $S_1$, $S_2$, …, $S_n$. So from the point of view of the buyer, the sequence of cash flows from the swap is $S_1 – FS_0(n, T)$, $S_2 – FS_0(n, T)$, …, $S_n – FS_0(n, T)$. 

A swap is in some sense a series of forward contracts, specifically a set of contracts expiring at various times in which one party agrees to make a fixed payment and receive a variable payment. 

A forward transaction that starts with a nonzero value is called an `off-market forward`. Each implicit forward contract is said to be off-market, because it is created at the swap price, not the appropriate forward price, which would be the price created in the forward market.

Suppose we buy an asset that pays the amounts S1, S2, …, Sn at times t1, t2, …, tn. These are unknown amounts. A simple example would be a floating-rate bond for which the S values represent the coupons that are unknown at the start but ultimately are determined by the evolution of interest rates. Then suppose we finance the purchase of that asset by borrowing money that we promise to repay with equal fixed payments of FS0(T). That strategy replicates the swap. As you have already learned, replication is the key to pricing.

**Valuation of the swap during its life again appeals to replication and the principle of no arbitrage.** The principal of replication articulates that the valuation of a swap is the present value of all the net cash flow payments from the swap, not simply the present value of the fixed payments of the swap or the present value of the underlying at the end of the contract.

## Pricing and Valuation: Options
Options are contingent claims. Call options represent the right to buy, while put options represent the right to sell.

Option prices can be significantly influenced by their style:
- European-style options can only be exercised at the expiration date.
- American-style options can be exercised any time up to the expiration date.

The formulas covered in this section use similar notation to what has been used in previous sections on derivatives:

$T$ is the expiration time

$X$ is the exercise price (i.e., strike price)

$S_0$ is the spot price of the underlying asset at initiation

$S_T$ is the spot price of the underlying asset at expiration

$S_t$ is the spot price of the underlying asset at some time between initiation and expiration

$c$ is the value of a European call

$C$ is the value of an American call

$p$ is the value of a European put

$P$ is the value of an American put

### Lower Limits for Prices of European Options
#### Value of European Option at Expiration
An investor who purchases a European call option has the right to buy an asset worth $S_T$ at a price of $X$ at expiration time $T$. Clearly, this option will only be exercised if the strike price is less than the price that would be paid to acquire the asset on the open market. Otherwise, the option will be worthless and the investor will walk away.

Therefore, the value of the European call option at expiration will either be zero or the difference between the asset price and the strike price:
$$c_T=max(0,S_T−X)$$

The valuation of a European put option is similar. The holder will either choose to sell the asset if it is trading for less than the strike price or walk away.
$$p_T=max(0,X−S_T)$$

#### Effect of Value of the Underlying
Clearly, the values of both call and put options are dependent on the price of the underlying at expiry, which cannot be known in advance. Call option prices are directly related to the price of the underlying asset, while the relationship is inverse for put options.
#### Effect of the Exercise Price
An option can be described in terms of moneyness based on the difference between its `strike price` and the `current price` of the underlying, $S_t$.

- If the underlying is trading above the exercise price ($S_t > X$), a call option is `in the money` and a put option with the same strike price is `out of the money`.
- If the underlying is trading below the exercise price ($S_t < X$), a call option is `out of the money` and a put option with the same strike price is `in the money`.
- if $X=S_t$, both call and put options are described as being `at the money`.

This means the relationship between the option value and exercise price is `inverse` for calls and `direct` for puts.
#### Effect of Time to Expiration
In general, options are more valuable if the time to expiration is longer. This is true because options are one-sided (i.e., the payoff can never be negative). If an option is currently out of the money, it is more likely to finish in the money if it expires in one month rather than one day.

However, this rule does not always hold for European put options. Recall that **the best possible outcome for the owner of a European put option is if the price of the underlying is zero**. If the underlying is trading at zero before expiry, then the put holder would prefer to receive the maximum payoff immediately rather than having to wait until expiry. Even if the underlying is worthless at expiry, the present value of a future payment is less than receiving the same payment immediately.

The relationship between option value and time to expiry is direct for calls. It is usually direct for puts as well, but it may be inverse if the risk-free rate is higher, the option is deeper in the money, or the time to expiry is longer.
#### Effect of the Risk-Free Rate of Interest
By owning a call option, an investor can gain from an increase in the price of the underlying without having to pay full price for it today. An increase in the risk-free rate reduces the present value of the call owner's (potential) future payment. For put option holders, a higher risk-free rate reduces the present value of the payment that they would receive by exercising their right to sell.

This implies that the value of a European call is directly related to the risk-free rate, while the value of a European put is inversely related.
#### Effect of Volatility of the Underlying
Generally, investors view volatility as a bad thing. However, for the holder of an option, volatility is good. This also occurs because options have one-sided payoff profiles. Movement of the underlying security price in the right direction can add value to the option payoff, while adverse price movement can never make the payoff negative. Therefore, call and put option values are directly related to the volatility of the underlying asset price.

Volatility and time are the sources of an option's time value, which is the difference between an option's price and its intrinsic value. 

*For example, if a stock is trading for \\$40, a 90-day call option with a \\$35 strike price will have an intrinsic value of \\$5 because it grants the right to purchase the stock for $5 less than its market value. However, this option would sell for \\$5.70, of which \\$5 is the intrinsic value and \\$0.70 is the time value.*

An option's time value approaches zero as it moves closer to maturity. This process is known as time value decay.
#### Effect of Payments on the Underlying and the Cost of Carry
Benefits from an underlying asset (e.g., dividends from stocks, interest from bonds, convenience yields from commodities) accrue to the current owner. This is bad for the holder of a call option, who will not own the asset until later (if ever), but good for the holder of a put option.

Of course, asset owners also have to incur carrying costs (e.g., storage). By holding call options, investors can gain from rising asset prices without incurring the carrying costs.

All else equal, benefits from underlying assets reduce the value of calls and increase the value of puts. Carrying costs have the opposite effect.
#### Lowest Prices of Calls and Puts
The minimum price for both European calls and puts can be derived using arbitrage principles. A portfolio using the risk-free asset and the underlying security that has a payoff never greater than the option must have a current value less than the option. That leads to the following two minimum option values

$$c_0≥max[0,S_0−\frac{X}{(1+r)^T}]$$
$$p_0≥max[0,\frac{X}{(1+r)^T}−S_0]$$

### Put-Call Parity
`Put-call parity` is an essential concept every trader should understand about options. It is derived by comparing the payoffs of two portfolio strategies – `protective put` and `fiduciary call`.

1. A `protective put` is composed of a long underlying asset $S$ plus a put option $p$.
2. A `fiduciary call` is composed of a risk-free asset that has a value of $X$ at time $T$ plus a call option $c$.

The payoff of both portfolios at time $T$ are identical regardless of the ending underlying asset price, which is illustrated in the table below.

**Since the payoffs are identical for the two portfolios, their values must also be equivalent. This is put-call parity.**
$$S_0+p_0=c_0+\frac{X}{(1+r)^T}$$

It must hold true or arbitrage possibilities will exist. The equation can easily be rearranged to solve for any of the four items.

$$p_0=c_0−S_0+\frac{X}{(1+r)^T}$$
$$c_0=S_0+p_0−\frac{X}{(1+r)^T}$$
$$S_0=c_0−p_0+\frac{X}{(1+r)^T}$$
$$\frac{X}{(1+r)^T}=S_0+p_0−c_0$$

### Binomial Valuation of Options
The payoff of an option can be determined relatively easily if the value of the underlying is known. An option pricing model then just requires a model of future underlying prices. One type of process is called the `binomial model`. It assumes the underlying can either go up by a factor of $u$ or down by a factor of $d$ during each small time interval. So after the first time interval, the underlying asset will move from a price of $S_0$ to $S_0u$ or $S_0d$. This is shown for a call option in the table below.

The volatility of the underlying drives the difference between the upper and lower values of the underlying, consequently the upper and lower values of the option.

A portfolio could be created at each node by buying a portion of the underlying asset $S$ and selling one call option $c$. The portfolio would have an initial value of $V_0=hS_0−c_0$. After one time period, the portfolio value would be:

$hS^+_1−c^+_1$ or $hS^−_1−c^−_1$

If these two potential values are set equal to each other, the portfolio would be risk-free by definition. Then h could be calculated, as shown below.

$hS^+_1−c^+_1=hS^−_1−c^−_1$

$h=\frac{c^+_1−c^−_1}{S^+_1−S^−_1}$

Since the portfolio is risk-free, it must grow at the risk-free rate of $r$. This allows the call price to be calculated using the pseudo (or risk-neutral) probability $π$.

$$c_0=\frac{πc^+_1+(1−π)c^−_1}{1+r}$$

$$π=\frac{1+r−d}{u−d}$$
