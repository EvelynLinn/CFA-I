# Basics of Portfolio Planning and Construction

## Portfolio Planning
### Investment Policy Statement (IPS) and its Major Components
Portfolio planning is done before constructing a portfolio. The `investment policy statement (IPS)` is the written document that governs this process. It serves as a reference manual for helping clients achieve their investment objectives. An IPS should be reviewed regularly to ensure that it remains consistent with the client's current circumstances.

Depending on circumstances, a written IPS or its equivalent may be required by law or regulation and a written IPS is certainly consistent with best practices. The mere fact that a written IPS is prepared for a client, however, does not ensure that risk and return objectives will in fact be achieved.

A general framework for an IPS includes the following components:
1. Introduction: This section describes the client.
2. Statement of purpose: This section states the purpose of the IPS.
3. Statement of duties and responsibilities: This section details the duties and responsibilities of the client, the custodian of the client’s assets, and the investment managers. 
4. Procedures: This section explains the steps to take to keep the IPS current and the procedures to follow to respond to various contingencies.
5. Investment objectives: This section explains the client’s objectives in investing.
6. Investment constraints: This section presents the factors that constrain the client in seeking to achieve the investment objectives.
7. Investment guidelines: This section provides information about how policy should be executed (e.g., on the permissible use of leverage and derivatives) and on specific types of assets excluded from investment, if any.
8. Evaluation and review: This section provides guidance on obtaining feedback on investment results.
9. Appendices: 
  - (A) Strategic Asset Allocation 
  - (B) Rebalancing Policy. 
  - Many investors specify a `strategic asset allocation (SAA)`, also known as the `policy portfolio`, which is the baseline allocation of portfolio assets to asset classes in view of the investor’s investment objectives and the investor’s policy with respect to `rebalancing` asset class weights. This SAA may include a s`tatement of policy` concerning hedging risks such as currency risk and interest rate risk.

### Risk Objectives
The most relevant components of the IPS are investment objectives (i.e., risks and returns) and investment constraints (e.g., liquidity, time horizon, taxes). 

A quantitative risk objective may be stated in absolute or relative terms. An absolute risk objective may be to require a 95% probability of not experiencing negative returns in a given year. Measures of absolute risk include 
- standard deviation
- variance
- value at risk (`VAR`)

A relative risk objective references a benchmark. An example is a 3% maximum `tracking error` (or `tracking risk`, the standard deviation of the differences between a portfolio’s returns and its benchmark’s returns) relative to a market index. Institutional clients with specific obligations may define risk relative to a liability benchmark. For example, a DB pension plan may express its risk objective as a 95% probability that the portfolio value will be within 2% of the projected benefit obligation.

    Assuming returns follow a normal distribution, an expected tracking risk of 2 percent would imply a return within 4 percent of the index return approximately 95 percent of the time. Remember that tracking risk is stated as a one standard deviation measure.

An assessment of a client's risk tolerance should consider both ability and willingness to accept risk. The determination of ability to take risk is based on largely objective criteria, such as time horizon and level of wealth. In general, longer time horizons and higher levels of wealth (relative to liquidity needs) indicate a greater ability to accept risk.

A client's willingness to take risk is based on much more subjective analysis. Academic researchers have developed various questionnaires that are designed to reveal respondents' risk tolerance based on their answers to questions about, for example, keeping money in a bank account rather than investing it in the stock market.

When there is a conflict between ability and willingness to take risk, a resolution is required. 
- In the case of a client with high ability and low willingness, the manager may attempt to educate the client about the opportunity cost of a low-risk portfolio. However, as a general rule, these efforts should not include attempts to modify the client’s personality. 
- The overall assessment of risk tolerance should be based on the lower of ability or willingness.

### Return Objectives
Absolute return objectives (e.g., 8% annually) can be stated in either **real or nominal** terms. Alternatively, investors may prefer a relative return objective, such as Libor + 400 basis points or greater than the average of a peer group of managers. Return objectives may be pre-tax or after-tax, but after-tax returns are what ultimately matter for taxable investors.
### IPS Constraints
#### Liquidity
`Liquidity needs` are withdrawals from the investment portfolio. For an individual investor, a down payment for a home would be a major liquidity need. An endowment's liquidity needs are ongoing spending commitments. Expected withdrawals should be documented in the IPS and the portfolio should be managed to ensure that there are sufficient liquid assets to meet these needs.
#### Time Horizon
The time horizon could be the period before withdrawals or before the client’s circumstances are likely to change. Investors with shorter time horizons should avoid holding significant amounts of illiquid assets.
#### Tax Concerns
Portfolio management decisions should account for the client's taxable status. Tax-exempt investors (e.g., pension plans) are unlikely to benefit from a large allocation to tax-free municipal bonds. For taxable investors, consideration should be given to different tax treatment of various types of income (e.g., dividends, capital gains).
#### Legal and Regulatory Factors
Institutional investors are often subject to minimum or maximum allocations to certain asset classes. There may be legal restrictions on foreign investments or the use of derivatives. `Self-investment limits` prohibit pension plans from overinvesting in the sponsor's securities. Individual investors who are corporate insiders may be subject to restrictions on trading their company's shares during quiet periods before material information has been made public. Any such restrictions should be documented in the IPS.
#### Unique Circumstances
Clients may self-imposed constraints based on, for example, ethical or religious beliefs. Investors may choose to avoid certain stocks or sectors by applying `negative screening` (also known as `exclusionary screening)`. 

Alternatively, a `best-in-class approach` seeks to identify top-performing companies according to environmental, social, and governance (ESG) metrics.

`Thematic investing` attempts to profit from important trends and structural changes, such as the transition to alternative sources of power generation. `Impact investing` is similar, but managers primarily consider the expected environmental or social benefits when selecting investments.

ESG integration, which involves adding an ESG perspective to the traditional process of investment analysis, is increasingly popular. However, investors and managers should be aware of potential **data quality** issues. There are no universal standards for ESG reporting, which is typically voluntary, and companies tend to limit their disclosures to the information that is most favorable to them.

If the manager is only investing a portion of the client’s total wealth, the investments section should consider the outside assets held by the client. For example, clients may want to exclude securities of their employers to limit the concentration of human and financial capital.
### Gathering Client Information
Investment advisors should know their clients. This is true even in advisory relationships. 

Good record keeping is also essential.

## Portfolio Construction and Capital Market Expectations
A `strategic asset allocation` is often the first step after completing the IPS. **The focus should be on systematic risk to make sure it is consistent with the client’s risk and return objectives.**
### Capital Market Expectations
`Capital market expectations` are the risk and return prospects of asset classes. It usually includes expected returns, standard deviations, and correlations.
### Strategic Asset Allocation
The major asset classes are:
- cash
- equities
- bonds
- real estate. 

Alternative investments include:
- private equity
- hedge funds
- commodities. 

The asset classes could be further subdivided, such as government and corporate bonds. Each asset class should be relatively homogenous.

The strategic asset allocation combines the constraints and objectives of the IPS with long-term capital market expectations. The risk-return profile is improved by adding asset classes with low correlations.

Investors’ **risk and return objectives** can be described as a utility function, in which utility increases with higher expected returns and lower risk.  A `utility function` such as the one below could be used to find the optimal asset allocation.

$$U_p=E(R_p)−λσ^2_p$$
<div style="text-align:right"> Equation 1 </div>

where

$U_p$	=	the investor’s expected utility from the portfolio

$E(R_p)$	=	the expected return of the portfolio

$σ_p$	=	the standard deviation of returns of the portfolio

$\lambda$	=	a measure of the investor’s risk aversion

The **optimal portfolio** occurs at the point of tangency between the indifference curve and the efficient frontier. 

    This utility function expresses a positive relationship between utility and expected portfolio return (i.e., higher expected return increases utility, all else equal) and a negative relationship between utility and volatility of portfolio return as measured by the variance of portfolio returns. The stronger the negative relationship, the greater the investor’s risk aversion. The portfolio is understood to represent a particular asset allocation. The asset allocation providing the highest expected utility is the one that is optimal for the investor given his or her risk aversion.
    
    For different values of Up, a line can be plotted that links those combinations of risk and expected return that produces that level of utility: an indifference curve. An investor would attain equal utility from all risk/return combinations on that curve.
    
Capital market expectations, specified in asset classes’ expected returns, standard deviations of return, and correlations, translate into an efficient frontier of portfolios. A multi-asset class portfolio’s expected return is given by

$$E(R_p)=\sum_{i=1}^n{w_iE(R_i)}$$
<div style="text-align:right"> Equation 2 </div>

where $w_i$ equals the weight of asset class $i$ in the portfolio, and its risk is given by

$$σ_p=\sqrt{\sum_{i=1}^n\sum_{j=1}^n{w_{p,i}w_{p,j}Cov(R_i,R_j)}}$$
<div style="text-align:right"> Equation 3 </div>
            
The covariance between the returns on asset classes i and j is given by the product of the correlation between the two asset classes and their standard deviations of return:

$$Cov(R_i,R_j) = ρ_{i,j}σ_iσ_j$$
<div style="text-align:right"> Equation 4</div>

where

$Cov(R_i,R_j)$	=	the covariance between the return of asset classes $i$ and $j$

$ρ_{i,j}$	=	the correlation between the returns of asset classes $i$ and $j$
### Steps Toward an Actual Portfolio
`Risk budgeting` involves deciding the amount of risk to assume in a portfolio and subdividing the risk over the return sources. The IPS defined the amount of risk, so the next step is subdividing the risk.

The risk can be subdivided into:
1. strategic asset allocation
2. tactical asset allocation
3. security selection

- **Strategic asset allocation** defines exposure to systematic risk.
- **Tactical asset allocation** Tactical asset allocation allows actual asset allocation to deviate from that of the strategic asset allocation (policy portfolio) of the IPS. Tactical asset allocation attempts to take advantage of temporary dislocations from the market conditions and assumptions that drove the policy portfolio decision. *(It seeks to add return by varying the weights between asset classes based on forecasts. Tactical asset allocation involves `overweighting` or `underweighting` specific asset classes, securities, or sectors based on short-term capital market expectations. Such deviations from the systematic risk exposure implied by a portfolio's policy weights create `non-systematic risk exposure` that would not be present if the optimal portfolio weights had been maintained.)*
- **Security selection** ibs the process of selecting individual securities; typically, security selection has the objective of generating superior risk-adjusted returns relative to a portfolio’s benchmark.

The formal definition of the `rebalancing` policy is the set of rules that guide the process of restoring a portfolio's original exposures to systematic risk factors. This may be done periodically according to a schedule or whenever asset class weights move outside of specified corridors.

Tactical asset allocation and security selection involve active management. Value added depends on the efficiency of the market and the skill of the manager.

Even passive strategies require some trading for rebalancing. This is needed due to drift and changing capital market expectations.

The `core-satellite approach` involves allocating the majority of a portfolio's assets to passive or low active investments (the "core") and placing the remaining funds in actively managed "`satellite" accounts`.
### Alternative Portfolio Organizing Principles
The proliferation of `exchange-traded funds (ETFs)` and algorithm-based robo-advisors have made it easier for investors to build portfolios with low-cost, liquid exposure to a broad range of asset classes.

Advocates of `risk-parity investing` call for asset classes to be weighted based on their contribution to a portfolio's overall risk. They argue that high allocations to equities understate the actual level of risk. Portfolios based on `risk-parity principles` have performed well since the global financial crisis of 2007 - 2009, although critics claim that this can be explained by persistently low interest rates that have enhanced fixed-income returns.
### ESG Considerations in Portfolio Planning and Construction
Environmental, social, and governance (ESG) issues are becoming an increasingly significant consideration in the portfolio construction and planning process. Investors may choose to screen out companies that do not meet specific ESG criteria or seek out best-in-class companies.

Investors who wish to pursue impact investing or thematic investing strategies may need to find specialist managers who are capable of identifying and exploiting relevant opportunities.

Although advocates argue that portfolios benefit from screening out companies with significant exposure to ESG risks, empirical evidence on this topic is mixed.

# Introduction to Risk Management

## The Risk Management Process
Risk is always present because outcomes are uncertain. The focus in this reading is on economic and financial risks, which must be managed by all businesses and investors. Good risk management will increase the chance of a preferred outcome, but it will not guarantee it. Risk management is also appropriate for individuals, but many of them take a `haphazard approach`. A systematic approach would likely yield better results.

Risk is exposure to uncertainty. In business and finance, it is not practical, or even desirable, to avoid risk completely. Investors who hold only risk-free assets are unlikely to achieve their long-term return objectives. Instead of avoiding risk altogether, individuals and companies should accept risks in the areas in which they have the expertise and manage their exposure to risks in other areas.

`Risk exposure` is the quantification of sensitivity to a certain risk. For example, banks have assets and liabilities that are sensitive to interest rate risk. It is possible to measure a bank's exposure to interest rate risk by modeling the impact of a 1% change in the yield curve on its net income. Risk exposures are dynamic, so they must be continuously monitored.

`Risk management` is the process by which an organization or individual defines the level of risk to be taken, measures the level of risk being taken, and adjusts the latter toward the former, with the goal of maximizing the company’s or portfolio’s value or the individual’s overall satisfaction, or utility.

The risk management process involves setting an optimal level of risk exposure, measuring the actual level of risk exposure, and making any necessary adjustments to reach the target level. The objective of risk management is to minimize the uncertainty of the impact of unpredictable events.

A risk management framework is the infrastructure, process, and analytics needed to effectively manage an organization's risk exposures. It should include the following key factors:
- **Risk governance**: This is the top-level system of structures, rights, and obligations. It is normally done at the board level, led by a risk management committee that is tasked with overseeing the company's risk exposures and providing an enterprise-wide perspective on risk management. (sets overall context)
- **Risk identification and measurement**: This makes up the main quantitative core elements of risk management. All of an organization's potential risk exposures should be assessed qualitatively and quantified as accurately as possible. Risk drivers should be identified to help model changes in risk exposures under various scenarios, including periods of extreme market stress.
- **`Risk infrastructure`**: This includes the people and systems needed to **assess (risk profile) and quantify** risk exposures. *(Improvements in technology have allowed for greater sophistication and precision in modeling risk. The appropriate amount of risk infrastructure will vary among companies depending on their size and the nature of their activities.)*
- **Policies and processes**: The overall vision for risk management is established by the risk management committee. This vision is implemented with policies and processes that govern day-to-day operations. Risk management policies and processes should be integrated into business activities.
- **Risk monitoring, mitigation, and management**: This is the most important part of the risk framework and also the most difficult. It is an active process that must be continuously reviewed.
- **Communications**: Relevant information about risk management must be clearly communicated throughout an organization. The risk management committee can give better guidance if it is part of a feedback loop that includes managers and employees.
- **Strategic analysis or integration**: Good risk management is key to increasing a company's value. A well-integrated risk management strategy can help identify activities that add value as well as those that may be destroying value.

## Risk Governance 
### An Enterprise View
Risk governance is a top-down process that directs risk management activities to align with and support the overall enterprise. This identifies the risk appetite or risk tolerance abilities of a company, meaning which risks are acceptable, and which risks are to be mitigated. It also provides guidance on the worst losses that could be tolerated in various scenarios.

Responsibilities of governing body (i.e., risk management committee) include:
1. Providing risk oversight
2. Determining organizational goals, direction, and priorities
3. Specifying `risk appetite or tolerance` (i.e., which risks and levels of exposure are acceptable)

The governing body must tie the organizational goals and risk framework together. Successfully implementing risk management strategies requires a strong commitment from those in the highest levels of an organization.

Governing bodies must strike an appropriate balance between giving clear guidance and giving managers sufficient leeway to use their judgment in implementing a risk management strategy. *(It is imperative the discussions about organizational risk appetite take place before a crisis. Too often, these discussions do not take place until after a crisis has already started. "`Red zones`" must be established that clearly define unacceptable risks and levels of exposure.)*

The `risk management committee` is a part of the risk governance structure at the operational level. A risk management committee should take a `firm-wide perspective`. *(For example, a pension fund manager cannot focus on the assets without considering factors such as the nature and timing of liabilities or the sponsor's ability and willingness to make future contributions.)*

`Enterprise-focused risk management (ERM)`, which takes a holistic view of the firm, is more likely to add value than narrower, less integrated approaches to risk management. ERM can be applied to individuals as well. For example, investors can factor human capital considerations into investment decisions.

Risk management committees foster good risk governance. It focuses on operational level activities. It is also wise to appoint a `chief risk officer (CRO)` who is responsible for developing and implementing an enterprise-wide risk management framework. Ideally, a CRO should be involved in decisions about the company's strategic direction.
### Risk Tolerance
The most important function of a governing body is to determine an appropriate level of risk tolerance. This is the process of identifying how much an organization is willing to lose by accepting exposure to certain risks and the level of opportunity costs that it is willing to tolerate by limiting its exposure to others.

An analysis of risk tolerance should seek to identify the following:
- `Internal shortfalls` that would result in failure to achieve critically important objectives. Specific metrics may include the percentage drop in revenue that would trigger debt covenants or the amount of cash flow needed to fund key capital projects.
- Risk drivers, or `external uncertainties` to which the organization is exposed (e.g., exchange rates, commodity prices, interest rates). (Risks (and risk drivers) arise from fundamental factors in macroeconomies and industries.)

While there is no specific formula for determining risk tolerance, there are a variety of quantitative methods that can be used to inform these decisions. 
- A company’s goals, its expertise in certain areas, and its strategies will help a board determine which risks the company may pursue and with how much intensity. 
- The ability of a company to respond dynamically to adverse events may allow for a higher risk tolerance. 
- The amount of loss a company can sustain without impairing its status as a going concern should factor into its risk tolerance. 
- The competitive landscape matters because both the board’s and investors’ expectations are usually developed in the context of how a company is positioned in its industry. 
- The government and regulatory landscape is important too, both in their `ex ante` demands on how companies approach risk and in the likely `ex post` reaction in the event of disasters. 

    *Quantitative analyses such as scenario analysis, economic models, and sensitivities to macro risk drivers might be used to assess where a board’s zone of comfort is bounded.*


- Factors that should not be allowed to influence assessments of risk tolerance include short-term reporting pressures, company size, and management compensation. (Personal motivations, beliefs, and agendas of board members (the agency problem); company size; whether the market environment seems stable; short-term pressures; and management compensation often affect risk tolerance in ways that might not be in line with the owners’ best interests)

If a company has the ability to adapt quickly to `adverse events` may allow for a higher risk tolerance. There are other factors, such as beliefs of board members and a stable market environment, which may but should not affect risk tolerance.

Having a defined risk tolerance helps to frame decisions in the context of a strategic tradeoff between risk and return. Enterprises should not just focus on high returns, which can be derived from leverage or by pursuing unjustifiably risky ventures.
### Risk Budgeting
Risk budgeting is an effort to quantify and allocate allowable risk for **both business and portfolio management**. It is the process of implementing risk tolerance in the everyday decisions that affect a company's actual risk exposures. Although a desirable element of risk governance, it cuts across the entire risk management framework, providing a focal point for each of the facets of risk management.

Risk budgeting can be `multi-dimensional` and complex or `single-dimensional` and simple. 
- Single-dimension measures include 
  - standard deviation
  - beta
  - value at risk
  - scenario loss
- More complex risk budgeting practices include factor approaches that implement tilts toward certain types of stocks (e.g., value vs. growth) while maintaining a target level of overall equity market risk. 
  - One popular approach evaluates risks by their underlying risk classes, such as equity, fixed income, commodity risk, and so on, and then **allocates investments by their risk class**. 
  - Also common are `risk factor approaches` to risk budgeting, in which exposure to various factors is used to attempt to capture associated risk premiums. An example would be to budget an allocation to give greater emphasis to value stocks based on the belief that they may provide a higher risk-adjusted return than growth stocks. 
  - This tactic might be layered over a strategic budget with a certain “beta” as the overall equity risk, supplemented with value and additional factor tilts specified up to some level.

Many organizations and investors end up with sub-optimal risk profiles simply because they fail to implement a proper risk budget. For example, many individual investors hold more of their employer's stock than they should because they do not consider the implications of concentrating their human and financial capital. Financial planners can help identify and adjust these allocations.

Adopting a risk budget encourages organizations and investors to approach decisions with the objective of **maximizing returns per unit of overall risk or excess risk** relative to a benchmark.

## Identification of Risk 
Two general categories of risk are financial and non-financial. Financial risks usually stem from **changes in market prices and rates**. Non-financial risks arise from a variety of sources beyond the financial markets. 
### Financial Risk 
#### Market Risk
Market risk arises from movements in interest rates, stock prices, exchange rates, and commodity prices. These movements are driven by fundamental factors such as economic conditions, world events, or company-specific developments. Companies and investors tend to give most of their attention to managing market risk because it is so  visible.
#### Credit Risk (a.k.a. `default risk` or `counterparty risk`)
The risk one party does not meet its obligation to pay the counterparty to a contract such as a loan or an interest rate swap. 

Credit risk is driven by the same factors as market risk, but it is ultimately an `asset-specific risk`. While market prices can rebound quickly after a downturn, borrowers who default may find it difficult to access capital markets for many years.
#### Liquidity Risk (a.k.a `transaction cost risk`)
Liquidity risk is the risk of a downward adjustment when selling a financial asset. (The risk that a financial instrument cannot be purchased or sold without a significant concession in price due to the size of the market.) While all securities have bid-ask spreads (`cost of liquidity`, it's a known cost, thus not considered as risk), liquidity risk relates to the possibility that spreads will widen significantly for certain types of assets during periods of `market stress`. *(Even under normal market conditions, it can be difficult to sell a large position (relative to average trading volume) without accepting a significant `price concession`.)* 

Sources of liquidity risk: 
- illiquid assets
- market liquidity varies over time and the market for specific assets may become less liquid
- as the size of a position increases, the cost and uncertainty associated with liquidating it will increase. *(In some extreme cases, there may be no price above zero at which the seller can sell the asset.)*

### Non-Financial Risk
#### Settlement Risk
Settlement risk is the risk of settling payments right before a default. For example, the long party of a forward contract could transfer funds but the short party may not transfer the asset. The root source of this risk is the timing of the payment process.
#### Legal Risk
Legal risk includes the risk of an entity being sued. It also includes the risk of the contract not being upheld by the legal system. Even weak cases can sometimes prevail in the courts.
#### Compliance Risk
Compliance risk includes regulatory, accounting, and tax risk. Laws and regulations are usually trying to catch up with the financial world. They may be changed in ways that were not anticipated.
#### Model Risk
It is the risk of a valuation error from improperly using a model. Model risk can be from using an inappropriate model or even using an appropriate model incorrectly.
##### Tail Risk
Often the distribution tail is fatter than implied by the probability distribution used to model the process. This commonly occurs when it is assumed that returns will be normally distributed and insufficient consideration is given to the possibility of extreme negative returns. Tail risk is known to be especially severe for the normal distribution, which tends to be overused in modeling. 

Ignoring the existence of fat tails in returns results in the situation in which market risk is often considered and dealt with in an oversimplified fashion.
#### Operational Risk
Operational risk is the risk from people and processes used by companies to produce the output. This includes both external risks (e.g., hackers) and internal risk (e.g., employee theft). Another example is rogue employees such as traders who execute unauthorized or illegal transactions that result in significant losses. Business interruptions are key for companies such as airlines.
#### Solvency Risk
Solvency risk is the risk the entity does not survive because it lacks the cash needed to fund day-to-day operations. This risk was highlighted during the financial crisis of 2008. Solvency risk can be mitigated by maintaining stable sources of `financing and cash reserves`.
#### Risks Unique to Individuals
Individuals face non-financial risks to their property (e.g., theft, damage) and their health. Financial planning should account for
- the risk of dying relatively young (mortality risk), as well as 
- the possibility of outliving one's assets (longevity risk), which is an increasing concern with an aging population. 

Just as companies face the risk of rogue employees, individual investors may be cheated by unethical financial advisors. Like organizations, individuals must take proactive steps to manage their risk exposures.


### Interactions Between Risks
Different types of risk often overlap and it is difficult to neatly categorize them. There is a considerable amount of interaction between risks. For example, credit risk and legal risk are closely linked for lenders because the recovery rate in the event of a bankruptcy will be determined by court proceedings. Credit risk is also heightened when markets are more volatile.

The interaction between different types of risk often produces `non-linear losses` that are worse than would be expected if each risk were considered in isolation. Adverse risk interaction occurs when losses caused by one type of risk exacerbate losses caused by another type of risk. *(For example, a levered portfolio may experience a 2% loss if asset prices fall by 1%. However, the portfolio value may drop by more than 20% in response to a 10% decrease in asset prices if positions need to be liquidated at a discount to generate cash for margin calls. Risk models must account for these interactions.)*

## Measuring and Modifying Risk 
### Drivers 
Risks are present because the future is unknown. Financial risks are largely attributable to economic risks. 

Often the risk of individual companies is thought to be `unsystematic`. However, seemingly unsystematic risk can turn into systematic risk. *(For example, poor credit risk management by one large bank can lead to a major crisis with widespread effects throughout the financial system.)*
### Metrics
Metrics refers to the **quantitative measure of risk exposure**. 
- Probability is the most basic metric. *(For example, a model may indicate a 5% likelihood of a company defaulting on its debt obligations. However, this is an insufficient measure of risk because it does not indicate the `magnitude` of expected losses in the event of a default.)*
- Standard deviation ($\sigma$)  is commonly used as a measure of dispersion, but it is only appropriate for normal distributions. 
- Beta ($\beta$), a measure of `systematic (market) risk`, is based on the assumption that investors can eliminate their security-specific risk by holding a diversified portfolio.

The following metrics are used to measure the risks of derivatives:
- Delta ($\delta$) (sensitivity of the derivative price to the underlying asset)
- Gamma ($\gamma$) (sensitivity of delta to the underlying asset) ($\delta$ and $\gamma$ are measures of the movement in an **option** price, given a movement in the underlying. )
- Vega (sensitivity of the derivative price to the price volatility of the underlying asset)
- Rho ($\rho$) (sensitivity of the derivative price to changes in interest rates)


- `Duration` is often used for fixed-income securities. It is analogous to delta because it measures the sensitivity of a fixed-income instrument (security/portfolio) to market interest rates.
- `Value at risk (VaR)` specifies the minimum loss over a given time period at a given probability. VaR measure includes three elements: 
  - a probability, 
  - a time period, 
  - a **minimum** possible loss stated in units of currency. 
  
  *(For example, a given bank could be expected to lose at least 2 million USD in one day 5\% of the time. Unfortunately, there are many ways to model the loss, so diverse estimates for VaR could occur for a given company. **VaR is based on an assumed probability distribution (usually the normal distribution) that may not reflect reality.**)*

    VaR does not tell the maximum loss. It only provides the loss estimate at the $x^{th}$ percentile. 
- `Conditional VaR (CVaR)` is the average loss that exceeds the VaR, which is a better measure of the tail risk. Scenario analysis and stress testing can be used to supplement VaR.
- Two measures in particular that are often used to complement VaR are `scenario analysis` and `stress testing` (A specific type of scenario analysis that estimates losses in rare and extremely unfavorable combinations of events or scenarios).
- Credit risk measurements have historically relied on ratings from agencies like Moody’s and S&P. On their own, lenders still check liquidity, profitability, and leverage. Credit events are rare for particular entities, making it tough to get a meaningful history that can be used to estimate future events. A `credit default swap (CDS)` on an issuing company has an observable price that acts as a signal to a bondholder of the risk cost of a default. Put options, exotic options, insurance contracts, and other financial instruments may contain valuable signals of the cost of rare adverse events, or at least the price of hedging them.
- Operational risk is very difficult to measure. For example, what is the likelihood of hackers breaking into a retailer’s credit card database?

### Methods of Risk Modification 
Actual risk exposure should be aligned with the acceptable risk. Risk management conversations often focus on reducing or hedging risks, but some situations call for increased exposure to certain risks. Four broad categories of risk modification are discussed below.
#### Prevention, Avoidance
Avoiding a risk altogether is certainly a safe option, but some risks may be unavoidable. Most risk avoidance has trade-offs (e.g., never flying to avoid a crash). People take risks to get some personal upside and they may derive utility from risky activities such as smoking and skydiving. Corporate risk management committees may prohibit certain activities to avoid a particular risk that they deem unacceptable.
#### Acceptance: Self-Insurance and Diversification
Risk acceptance can be done with self-insurance and diversification. If it is too expensive to eliminate certain risks, individuals can effectively insure themselves by setting aside sufficient capital that can be accessed later if necessary. Self-insurance should be a conscious decision to accept certain risks that are consistent with a company's (or individual's) risk tolerance. Choosing to simply ignore risks that are beyond one's risk tolerance is not self-insurance, it is denial.

Diversification can be an effective method of managing risks that have been accepted. For example, every equity investor accepts market risk, but a portfolio that lays on the efficient frontier offers a better risk-adjusted return than a less diversified portfolio. Diversification, which is more accurately described as a risk mitigation strategy, is more effective if it is used in coordination with other risk management practices.
#### Transfer
Insurance policies, which have been around for at least 4,000 years, are the most common form of risk transfer. Insurers charge premiums in exchange for covering certain risks. These financial intermediaries can be profitable if the correlations between the risks that they have insured are low. If correlations are too high, insurers can transfer some of their risk to a reinsurer or issue catastrophe bonds that require investors to cover claims over a certain level.
- `Deductibles` are used to encourage caution from the insured by ensuring the policyholder has a financial stake in the risk. For example, drivers have a financial incentive to be more careful if they are responsible for covering a certain amount of repair costs in the event of an accident.
- `Surety bonds` are contracts in which the insurer accepts the risk of non-performance. For example, if a company orders a custom-built machine, it can use a surety bond to protect itself against the risk that the manufacturer will not meet its obligations. In exchange for a premium, the insurer will agree to compensate the company if the machine is unsatisfactory. 
- `Fidelity` bonds are similar contracts that companies use to protect themselves against losses due to employee dishonesty.

#### Shifting
Risk shifting changes the distribution of the risk outcome, often with derivatives. Derivatives, which are covered in great detail in other readings, can be classified into two categories: forward commitments and contingent obligations.

`Forward commitments` include 
- forwards
- futures
- swaps

They require no upfront premium and are an obligation for both parties. 

`Contingent claims` include 
- call and put options. 

They do require upfront premiums and the holder of the options has the right, but not the obligation, to exercise.
#### Choosing a Method for Modifying
There is no single “correct” method of risk modification. All of the methods covered here have trade-offs. In general, risk avoidance and prevention are recommended methods for managing risks that offer relatively little upside and potentially extreme losses. `Self-insurance` is a more feasible option for companies with significant free cash flows. Risk transfer may not be an option for managing risks that cannot be pooled. Risk shifting may not be possible for risks that are not covered by derivative contracts.

Risk acceptance is similar to `self-insurance`. An organization choosing to self-insure may set up a reserve fund to cover losses. 

Risk takers should identify risks that offer few rewards in light of potential costs and avoid those risks when possible. They should self-insure where it makes sense and diversify to the extent possible. They should consider insurance when risks can be pooled effectively if the cost of the insurance is less than the expected benefit. If derivatives are used, they must consider the trade-off of locking in outcomes with forward commitments versus the flexibility relative to cash cost of contingent claims, which can tailor the desired outcomes or payoffs by shifting the risk. Ultimately, the decision is always one of balancing costs against benefits while producing a risk profile that is consistent with the risk management objectives of the organization.

# Fintech in Investment Management

## Fintech and Big Data
`Fintech` is a technological innovation in financial services and products. This includes the applications of Big Data, artificial intelligence, and machine learning in finance.

Fintech began with the automation of routine tasks, but it has evolved into decision-making applications based on machine-learning logic.

Key areas of fintech include:
- Analysis of large datasets, such as from social media and sensor networks
- Analytical tools, including artificial intelligence techniques
- Automated trading, which can lead to more efficient and anonymous trading
- Automated advice, known as “robo-advisers,” that are cheaper than traditional advisers
- Financial record keeping, done using such means as `distributed ledger technology (DLT)`, which reduces the need for financial intermediaries

Datasets are growing in size and diversity. Big Data includes traditional data, such as stock exchanges and governments, and alternative data, including social media and sensor networks.

Big Data datasets have the characteristics of volume, velocity, and variety. Data volumes have grown from megabytes to petabytes. High velocity is possible because more data are now available in real-time. And the variety of Big Data includes unstructured data from sources such as social media, emails, and text messages.
### Sources of Big Data
Big Data sources include:
- financial markets
- businesses (e.g., corporate financials)
- governments (e.g., trade and economic data)
- individuals (e.g., credit card purchases, internet search logs, and social media posts)
- sensors (e.g., satellite imagery and traffic patterns)
- Internet of Things (e.g., data from “smart” buildings)

The three main sources of alternative data are generated by individuals, business processes, and sensors.

Data from individuals are often unstructured and formatted as text, video, photo, and audio. More data become available as individuals engage in more activities with electronic trails, such as on social media and web searches.

Business process data are more structured. They encompass data from direct sales and `corporate exhaust`, which include supply chain information and banking records. Retail point-of-sale scanner data are an example of corporate exhaust that is generated by business processes to provide a real-time indication of business performance.

Sensors on devices such as smartphones and satellites are the largest source of data. Sensor data include network arrangements known as the Internet of Things. These networks allow objects in the system (e.g., buildings, appliances, vehicles) to interact and share information.

Alternative data are used to identify factors that affect security prices, which can then be used to improve asset selection and trading. But investment professionals should be cautious about collecting personal information that is protected by regulations.
### Big Data Challenges
Challenges with Big Data include **quality, volume, and appropriateness**. The dataset could be biased or have outliers. The dataset must be sourced, cleansed, and organized before it can be used. Artificial intelligence and machine learning techniques are being used to sort this complex information.

## Advanced Analytical Tools: Artificial Intelligence and Machine Learning
Artificial intelligence (AI) enables computers to perform tasks that traditionally have required human intelligence. AI has been used by financial institutions since the 1980s, beginning with neural networks used to detect credit card fraud.

Machine learning (ML) algorithms are computer programs that learn how to complete tasks, improving with time as more data has become available. ML models are trained to map relationships between inputs and outputs. A training dataset is used to create the model, and other datasets are used to validate the model. Algorithms are not explicitly programmed, which can result in outcomes that are not easily understood.

Human judgment is needed to ensure data quality. Potential errors include overfitting, which occurs when an ML model learns the inputs and target dataset too well. Overfitting can cause ML models to treat noise in the data as true parameters. Underfitting can cause the ML model to treat true parameters as noise. These models may be too simplistic.
### Types of Machine Learning
Machine learning includes `supervised` and `unsupervised` learning.

In supervised learning, computers learn from labeled training data. The inputs and outputs are identified for the algorithm. This approach could be used to identify the best variable to forecast future stock returns.

With unsupervised learning, only the dataset is provided to the algorithm. The inputs and outputs are not labeled. This could be used to group companies into peer groups.

Deep learning utilizes neural networks to identify patterns with a multistage approach. This leads to an understanding of simple concepts that can be used to create more complex concepts. These algorithms are used in applications such as image and speech recognition.
### Data Science: Extracting Information from Big Data
Data science uses computer science and statistics to extract information from data.
#### Data Processing Methods
`Data capture` – Collecting data and transforming them to a usable format. Low-latency systems communicate high volumes of data with minimal delay, which is needed for automated trading.

`Data curation` – Cleaning data to ensure high quality.

`Data storage` – Recording, archiving, and accessing data.

`Search` – Locating specific information in large datasets.

`Transfer` – Moving data from their source or storage location to the analytical tool.
#### Data Visualization
Data visualization is how the data will be displayed and summarized in graphical form. `Multidimensional analysis` is often needed, which can be done by adding color and shapes to 3D charts. Tag clouds can be used to show the frequency of keywords in the data. Words that appear more often are in a larger font.

Common programming languages used in data science include Python, R, Java, C/C++, and Excel VBA. Databases used include SQL, SQL Lite, and NoSQL.

## Selected Applications of Fintech to Investment Management
### Text Analytics 
Text analytics use computer programs to analyze unstructured text- or voice-based datasets. This could be from data such as company filings, social media, and email. These methods can be used to identify predictors of market movements.
### Natural Language Processing
Natural language processing (NLP) focuses on interpreting human language. This is used in translation and speech recognition applications. It can also be used to monitor communications among employees to ensure compliance with policies.

NLP can be used to monitor more reports and earnings calls than is humanly possible. For example, NLP can be used to monitor analyst commentary regarding investment recommendations. This can be used to deduce the sentiment in the report, which could lead to rating changes later.
### Robo-Advisory Services
Robo-advisory services first appeared in 2008. These online platforms reduce the need for clients to interact directly with human financial advisors.

In the United States, robo-advisors must be registered as investment advisors. Many other countries also regulate them like other investment professionals.

Robo-advice is given based on a client's data, including their answers to a risk tolerance questionnaire. The services provided range from asset allocation to tax-loss harvesting. Most recommendations utilize passive investment approaches, creating a very low-cost alternative for clients.

Two types of wealth management services are common with robo-advice.
- Fully automated digital wealth managers: These provide a low-cost solution and often recommend a mix of exchange-traded funds.
- Advisor-assisted digital wealth managers: Virtual financial advisors can offer basic planning advice over the phone, overcoming some limitations with robo-advisors.

Robo-advisors are useful for investors with less wealth that cannot afford a personal advisor. But clients may not trust robo-advisors to make all the decisions. And robo-advisors may not be able to properly address complex situations.
### Risk Analysis
Regulators demand stress testing and risk assessment by investment firms. To monitor market activity in real-time, regulators must analyze substantial volumes of data, including quantitative and qualitative risk data. ML techniques can be used to analyze data quality. Portfolio risk management can be done with `scenario analysis`, which is computationally intense.
### Algorithmic Trading
Algorithmic trading is automated trading of financial instruments based on pre-defined rules. It is often used to break very large orders into smaller pieces. This type of trading increases the speed of execution and provides more anonymity. Algorithmic trading strategies may be executed using either `limit orders` to establish a maximum purchase price/minimum sale price or market orders that take the prevailing market price.

`High-frequency trading (HFT)` is a type of algorithmic trading that occurs in fractions of a second. This trading seeks to profit from intraday market mispricings.

## Distributed Ledger Technology
`Distributed ledger technology (DLT)` can be used to improve financial record keeping. Ownership of assets is created and exchanged on a peer-to-peer basis. DTL can improve accuracy and transparency. However, there are concerns about potential breaches of privacy because the technology is not fully secured. Additionally, DTL requires a massive amount of computing power to verify transactions.

A distributed ledger is shared by entities in a network. Each participant has a matching copy of the digital database. A consensus mechanism is used to ensure that all participants agree on a common state of the ledger. Transactions are validated as they occur and the ledger is updated accordingly. DTL networks use `cryptography` to verify participants and encrypt data to make them unusable for unauthorized parties.

`Smart contracts` can be created with DLT. These contracts are computer programs that will be executed if pre-specified terms are met. For example, collateral could get transferred automatically if a default occurs.

Blockchain is a type of digital ledger in which information is recorded sequentially within blocks that are linked (i.e., chained) using cryptography. New transactions are added only after they have been validated through a consensus mechanism. The process of updating a blockchain requires a substantial amount of computing power, which makes it difficult to manipulate historical data. These networks require broad participation so that no single entity can control the majority of `nodes`.
### Permissioned and Permissionless Networks
`Permissionless networks` allow any user to make a transaction. All users can see the transactions that exist on a blockchain. No centralized authority is needed to validate transactions. Because every node stores a copy of the database, it cannot be changed. Bitcoin is an example of an open, permissionless network.

`Permissioned networks` can restrict members from participating in some activities. For example, a network may allow certain participants to add transactions, while regulators would only be authorized to view transaction data.
### Applications of Distributed Ledger Technology to Investment Management
#### Cryptocurrencies
A cryptocurrency is an electronic currency that allows transactions between parties without an intermediary, such as a bank. Most cryptocurrencies use open DLT systems. Due to the high volatility that several cryptocurrencies have experienced, their effectiveness as a store of value has diminished. Many cryptocurrencies have experienced high levels of volatility due to a lack of clear fundamentals.

`Initial coin offerings (ICOs)` are used by companies to raise capital. Crypto tokens are sold to investors, who can use them to purchase future products by the issuer. Currently, ICOs are unregulated, which makes them attractive compared to traditional initial public offerings for companies seeking to raise capital quickly.

ICOs are unregulated, which significantly reduces issuance costs. However, authorities are considering whether and how to regulate ICOs following several incidents of fraudulent activity.
#### Tokenization
Physical assets, such as real estate, often require substantial effort to verify ownership. Tokenization is the process of representing ownership on a blockchain or distributed ledger. All historical activity will be on the same digital record.
#### Post-Trade Clearing and Settlement
Financial intermediaries confirm, clear, and settle transactions in financial securities markets. DLT can streamline this with near-real-time trade verification and reconciliation. This would reduce the need for third-party facilitation. The increased speed would reduce counterparty risk.
#### Compliance
Regulators are requiring more reporting and transparency. Currently, this is largely a manual process. DLT could potentially increase accuracy and transparency by providing regulators and firms with a near-real-time review of all transactions.
