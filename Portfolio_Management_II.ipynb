# Basics of Portfolio Planning and Construction

## Portfolio Planning
### Investment Policy Statement (IPS) and its Major Components
Portfolio planning is done before constructing a portfolio. The `investment policy statement (IPS)` is the written document that governs this process. It serves as a reference manual for helping clients achieve their investment objectives. An IPS should be reviewed regularly to ensure that it remains consistent with the client's current circumstances.

Depending on circumstances, a written IPS or its equivalent may be required by law or regulation and a written IPS is certainly consistent with best practices. The mere fact that a written IPS is prepared for a client, however, does not ensure that risk and return objectives will in fact be achieved.

A general framework for an IPS includes the following components:
1. Introduction: This section describes the client.
2. Statement of purpose: This section states the purpose of the IPS.
3. Statement of duties and responsibilities: This section details the duties and responsibilities of the client, the custodian of the client’s assets, and the investment managers. 
4. Procedures: This section explains the steps to take to keep the IPS current and the procedures to follow to respond to various contingencies.
5. Investment objectives: This section explains the client’s objectives in investing.
6. Investment constraints: This section presents the factors that constrain the client in seeking to achieve the investment objectives.
7. Investment guidelines: This section provides information about how policy should be executed (e.g., on the permissible use of leverage and derivatives) and on specific types of assets excluded from investment, if any.
8. Evaluation and review: This section provides guidance on obtaining feedback on investment results.
9. Appendices: 
  - (A) Strategic Asset Allocation 
  - (B) Rebalancing Policy. 
  - Many investors specify a `strategic asset allocation (SAA)`, also known as the `policy portfolio`, which is the baseline allocation of portfolio assets to asset classes in view of the investor’s investment objectives and the investor’s policy with respect to `rebalancing` asset class weights. This SAA may include a s`tatement of policy` concerning hedging risks such as currency risk and interest rate risk.

### Risk Objectives
The most relevant components of the IPS are investment objectives (i.e., risks and returns) and investment constraints (e.g., liquidity, time horizon, taxes). 

A quantitative risk objective may be stated in absolute or relative terms. An absolute risk objective may be to require a 95% probability of not experiencing negative returns in a given year. Measures of absolute risk include 
- standard deviation
- variance
- value at risk (`VAR`)

A relative risk objective references a benchmark. An example is a 3% maximum `tracking error` (or `tracking risk`, the standard deviation of the differences between a portfolio’s returns and its benchmark’s returns) relative to a market index. Institutional clients with specific obligations may define risk relative to a liability benchmark. For example, a DB pension plan may express its risk objective as a 95% probability that the portfolio value will be within 2% of the projected benefit obligation.

    Assuming returns follow a normal distribution, an expected tracking risk of 2 percent would imply a return within 4 percent of the index return approximately 95 percent of the time. Remember that tracking risk is stated as a one standard deviation measure.

An assessment of a client's risk tolerance should consider both ability and willingness to accept risk. The determination of ability to take risk is based on largely objective criteria, such as time horizon and level of wealth. In general, longer time horizons and higher levels of wealth (relative to liquidity needs) indicate a greater ability to accept risk.

A client's willingness to take risk is based on much more subjective analysis. Academic researchers have developed various questionnaires that are designed to reveal respondents' risk tolerance based on their answers to questions about, for example, keeping money in a bank account rather than investing it in the stock market.

When there is a conflict between ability and willingness to take risk, a resolution is required. 
- In the case of a client with high ability and low willingness, the manager may attempt to educate the client about the opportunity cost of a low-risk portfolio. However, as a general rule, these efforts should not include attempts to modify the client’s personality. 
- The overall assessment of risk tolerance should be based on the lower of ability or willingness.

### Return Objectives
Absolute return objectives (e.g., 8% annually) can be stated in either **real or nominal** terms. Alternatively, investors may prefer a relative return objective, such as Libor + 400 basis points or greater than the average of a peer group of managers. Return objectives may be pre-tax or after-tax, but after-tax returns are what ultimately matter for taxable investors.
### IPS Constraints
#### Liquidity
`Liquidity needs` are withdrawals from the investment portfolio. For an individual investor, a down payment for a home would be a major liquidity need. An endowment's liquidity needs are ongoing spending commitments. Expected withdrawals should be documented in the IPS and the portfolio should be managed to ensure that there are sufficient liquid assets to meet these needs.
#### Time Horizon
The time horizon could be the period before withdrawals or before the client’s circumstances are likely to change. Investors with shorter time horizons should avoid holding significant amounts of illiquid assets.
#### Tax Concerns
Portfolio management decisions should account for the client's taxable status. Tax-exempt investors (e.g., pension plans) are unlikely to benefit from a large allocation to tax-free municipal bonds. For taxable investors, consideration should be given to different tax treatment of various types of income (e.g., dividends, capital gains).
#### Legal and Regulatory Factors
Institutional investors are often subject to minimum or maximum allocations to certain asset classes. There may be legal restrictions on foreign investments or the use of derivatives. `Self-investment limits` prohibit pension plans from overinvesting in the sponsor's securities. Individual investors who are corporate insiders may be subject to restrictions on trading their company's shares during quiet periods before material information has been made public. Any such restrictions should be documented in the IPS.
#### Unique Circumstances
Clients may self-imposed constraints based on, for example, ethical or religious beliefs. Investors may choose to avoid certain stocks or sectors by applying `negative screening` (also known as `exclusionary screening)`. 

Alternatively, a `best-in-class approach` seeks to identify top-performing companies according to environmental, social, and governance (ESG) metrics.

`Thematic investing` attempts to profit from important trends and structural changes, such as the transition to alternative sources of power generation. `Impact investing` is similar, but managers primarily consider the expected environmental or social benefits when selecting investments.

ESG integration, which involves adding an ESG perspective to the traditional process of investment analysis, is increasingly popular. However, investors and managers should be aware of potential **data quality** issues. There are no universal standards for ESG reporting, which is typically voluntary, and companies tend to limit their disclosures to the information that is most favorable to them.

If the manager is only investing a portion of the client’s total wealth, the investments section should consider the outside assets held by the client. For example, clients may want to exclude securities of their employers to limit the concentration of human and financial capital.
### Gathering Client Information
Investment advisors should know their clients. This is true even in advisory relationships. 

Good record keeping is also essential.

## Portfolio Construction and Capital Market Expectations
A `strategic asset allocation` is often the first step after completing the IPS. **The focus should be on systematic risk to make sure it is consistent with the client’s risk and return objectives.**
### Capital Market Expectations
`Capital market expectations` are the risk and return prospects of asset classes. It usually includes expected returns, standard deviations, and correlations.
### Strategic Asset Allocation
The major asset classes are:
- cash
- equities
- bonds
- real estate. 

Alternative investments include:
- private equity
- hedge funds
- commodities. 

The asset classes could be further subdivided, such as government and corporate bonds. Each asset class should be relatively homogenous.

The strategic asset allocation combines the constraints and objectives of the IPS with long-term capital market expectations. The risk-return profile is improved by adding asset classes with low correlations.

Investors’ **risk and return objectives** can be described as a utility function, in which utility increases with higher expected returns and lower risk.  A `utility function` such as the one below could be used to find the optimal asset allocation.

$$U_p=E(R_p)−λσ^2_p$$
<div style="text-align:right"> Equation 1 </div>

where

$U_p$	=	the investor’s expected utility from the portfolio

$E(R_p)$	=	the expected return of the portfolio

$σ_p$	=	the standard deviation of returns of the portfolio

$\lambda$	=	a measure of the investor’s risk aversion

The **optimal portfolio** occurs at the point of tangency between the indifference curve and the efficient frontier. 

    This utility function expresses a positive relationship between utility and expected portfolio return (i.e., higher expected return increases utility, all else equal) and a negative relationship between utility and volatility of portfolio return as measured by the variance of portfolio returns. The stronger the negative relationship, the greater the investor’s risk aversion. The portfolio is understood to represent a particular asset allocation. The asset allocation providing the highest expected utility is the one that is optimal for the investor given his or her risk aversion.
    
    For different values of Up, a line can be plotted that links those combinations of risk and expected return that produces that level of utility: an indifference curve. An investor would attain equal utility from all risk/return combinations on that curve.
    
Capital market expectations, specified in asset classes’ expected returns, standard deviations of return, and correlations, translate into an efficient frontier of portfolios. A multi-asset class portfolio’s expected return is given by

$$E(R_p)=\sum_{i=1}^n{w_iE(R_i)}$$
<div style="text-align:right"> Equation 2 </div>

where $w_i$ equals the weight of asset class $i$ in the portfolio, and its risk is given by

$$σ_p=\sqrt{\sum_{i=1}^n\sum_{j=1}^n{w_{p,i}w_{p,j}Cov(R_i,R_j)}}$$
<div style="text-align:right"> Equation 3 </div>
            
The covariance between the returns on asset classes i and j is given by the product of the correlation between the two asset classes and their standard deviations of return:

$$Cov(R_i,R_j) = ρ_{i,j}σ_iσ_j$$
<div style="text-align:right"> Equation 4</div>

where

$Cov(R_i,R_j)$	=	the covariance between the return of asset classes $i$ and $j$

$ρ_{i,j}$	=	the correlation between the returns of asset classes $i$ and $j$
### Steps Toward an Actual Portfolio
`Risk budgeting` involves deciding the amount of risk to assume in a portfolio and subdividing the risk over the return sources. The IPS defined the amount of risk, so the next step is subdividing the risk.

The risk can be subdivided into:
1. strategic asset allocation
2. tactical asset allocation
3. security selection

- **Strategic asset allocation** defines exposure to systematic risk.
- **Tactical asset allocation** seeks to add return by varying the weights between asset classes based on forecasts. Tactical asset allocation involves overweighting or underweighting specific asset classes, securities, or sectors based on short-term capital market expectations. Such deviations from the systematic risk exposure implied by a portfolio's policy weights create non-systematic risk exposure that would not be present if the optimal portfolio weights had been maintained.
- **Security selection** can add return by picking securities with higher returns.

The formal definition of the `rebalancing` policy is the set of rules that guide the process of restoring a portfolio's original exposures to systematic risk factors. This may be done periodically according to a schedule or whenever asset class weights move outside of specified corridors.

Tactical asset allocation and security selection involve active management. Value added depends on the efficiency of the market and the skill of the manager.

Even passive strategies require some trading for rebalancing. This is needed due to drift and changing capital market expectations.

The `core-satellite approach` involves allocating the majority of a portfolio's assets to passive or low active investments (the "core") and placing the remaining funds in actively managed "`satellite" accounts`.
### Alternative Portfolio Organizing Principles
The proliferation of `exchange-traded funds (ETFs)` and algorithm-based robo-advisors have made it easier for investors to build portfolios with low-cost, liquid exposure to a broad range of asset classes.

Advocates of `risk-parity investing` call for asset classes to be weighted based on their contribution to a portfolio's overall risk. They argue that high allocations to equities understate the actual level of risk. Portfolios based on `risk-parity principles` have performed well since the global financial crisis of 2007 - 2009, although critics claim that this can be explained by persistently low interest rates that have enhanced fixed-income returns.
### ESG Considerations in Portfolio Planning and Construction
Environmental, social, and governance (ESG) issues are becoming an increasingly significant consideration in the portfolio construction and planning process. Investors may choose to screen out companies that do not meet specific ESG criteria or seek out best-in-class companies.

Investors who wish to pursue impact investing or thematic investing strategies may need to find specialist managers who are capable of identifying and exploiting relevant opportunities.

Although advocates argue that portfolios benefit from screening out companies with significant exposure to ESG risks, empirical evidence on this topic is mixed.
