# Understanding Fixed Income Risk and Return
The return on a fixed-rate bond is affected by many factors, the most important of which is the receipt of the interest and principal payments in the full amount and on the scheduled dates. Assuming no default, the return is also affected by changes in interest rates that affect coupon reinvestment and the price of the bond if it is sold before it matures. Measures of the price change can be derived from the mathematical relationship used to calculate the price of the bond. The first of these measures (`duration`) estimates the change in the price for a given change in interest rates. The second measure (`convexity`) improves on the duration estimate by taking into account the fact that the relationship between price and yield-to-maturity of a fixed-rate bond is not linear.

For a `discount bond` offers the investor a `“deficient” coupon rate`, or one below the market discount rate. The amortization of the discount in each period brings the return in line with the market discount rate as the bond’s carrying value is “`pulled to par`”.

For a `premium bond`, the coupon rate exceeds the market discount rate and the amortization of the premium adjusts the return to match the market discount rate. Through amortization, the bond’s carrying value reaches par value at maturity.

## Sources of Return
A fixed-rate bond has three sources of return:
1. Receipt of coupon payments and principal on scheduled dates
2. Reinvestment of coupon payments
3. Capital gains or losses

YTM is the rate that investors will earn under the following conditions:
1. The bond does not default
2. Coupons are reinvested at the YTM
3. The bond is held to maturity or sold on the constant yield price trajectory

The `constant yield price trajectory` is the path that a bond's price will follow as it approaches maturity if the YTM remains constant. For simplicity, consider the example of a 3-year, 5\% coupon bond with a par value of $100. Assuming an initial YTM of 7\%, the bond's initial price will be:

Note that this is a discount to par value because the bond's coupon rate is less than the yield required by investors. If the YTM remains constant at 7\%, the bond's price at the end of the next two year will be:

- End of Year 1:
$\frac{5}{1.07}+\frac{105}{1.07^2}=96.38$
- End of Year 2: 
$\frac{105}{1.07}=98.13$

Because the bond was issued at a discount, its price will increase over time as it is "pulled to par." 

A point on the trajectory represents the `carrying value` of the bond at that time. As shown in the graph below, a capital gain or loss will only be realized if the YTM changes, which would move the bond's price off its trajectory. 

$$capital\ gain\ or\ loss=sale\ price - carrying\ value$$

### Interest Rate Risk on Fixed-Rate Bonds
Changing interest rates do not have a consistent effect on all sources of a bond's return. Higher yields will reduce the bond's price, but increase the amount of reinvestment income. Lower yields will have the opposite effect. In other words, coupon reinvestment risk and market price risk offset each other.

Fixed-income investors holding the same bond can have different exposures to interest rate risk if their investment horizons differ. 

For investors with longer investment horizons, reinvestment risk is a larger concern than market price risk while investors with shorter investment horizons are more sensitive to market price risk.

Investors who intend to hold a bond until maturity do not need to be concerned with market price risk because, assuming no default, they will receive the bond's par value at maturity regardless of how much its price fluctuates during its lifetime.

## Bond Duration and Convexity
This section covers two commonly used measures of interest rate risk: `duration` and `convexity`. It distinguishes between risk measures based on changes in a bond’s own yield to maturity (yield duration and convexity) and those that affect the bond based on changes in a benchmark yield curve (curve duration and convexity).

Duration is a measure of a bond's price sensitivity to interest rate changes. (The duration of a bond measures the sensitivity of the bond’s full price (including accrued interest) to changes in the bond’s yield-to-maturity or, more generally, to changes in benchmark interest rates.) 

**<div style="text-align: center"> "How much would the bond price change immediately if interest rates changed?"</div>**

Duration estimates changes in the bond price assuming that variables other than the yield-to-maturity or benchmark rates are held constant. Most importantly, the time-to-maturity is unchanged. It represents the approximate holding period required to earn a return equal to a bond's yield-to-maturity (YTM) if interest rates experience a single change. If the bond is held for the duration period, an increase from reinvesting coupons is offset by a decrease in price if interest rates increase and a decrease from reinvesting coupons is offset by an increase in price if interest rates decrease.
#### Statistic Methods for Calculating Interest Rate Risk
There are multiple duration statistics, which fall into the following categories:
- `Yield duration` measures the sensitivity of a bond's price to its own YTM. Yield duration statistics include 
  - Macaulay duration, 
  - modified duration, 
  - money duration, and 
  - the price value of a basis point (PVBP).
- `Curve duration` (such as `effective duration`) measures sensitivity to changes in a benchmark yield curve, such as spot rates for zero-coupon government bonds of various maturities, government par curve. 
  - `Effective duration` is the most commonly used curve duration statistic.

### Macaulay Duration
`Macaulay duration` is named after the Canadian economist Frederick Macaulay. It is the **weighted average life of the cash flows, with the weights being the present value of the cash flows**. 

A general formula to calculate the Macaulay duration (MacDur) of a traditional fixed-rate bond.

$$MacDur=\frac{\frac{(1−t/T)×PMT}{(1+r)^{1−t/T}}+\frac{(2−t/T)×PMT}{(1+r)^{2−t/T}}+⋯+\frac{(N−t/T)×(PMT+FV)}{(1+r)^{N−t/T}}}{\frac{PMT}{(1+r)^{1−\frac{t}{T}}}+\frac{PMT}{(1+r)^{2−\frac{t}{T}}}+⋯+\frac{PMT+FV}{(1+r)^{N−\frac{t}{T}}}}$$

<div style="text-align: right"> Equation 1 </div>
where
$t$	=	the number of days from the last coupon payment to the settlement date

$T$	=	the number of days in the coupon period

$t/T$	=	the fraction of the coupon period that has gone by since the last payment

$PMT$	=	the coupon payment per period

$FV$	=	the future value paid at maturity, or the par value of the bond

$r$	=	the yield-to-maturity, or the market discount rate, per period

$N$	=	the number of evenly spaced periods to maturity as of the beginning of the current period

The denominator in Equation 1 is the full price ($PV^{Full}$) of the bond including accrued interest. It is the present value of the coupon interest and principal payments, with each cash flow discounted at YTM, $r$.
$$PV^{Full}=\frac{PMT}{(1+r)^{1−\frac{t}{T}}}+\frac{PMT}{(1+r)^{2−\frac{t}{T}}}+⋯+\frac{PMT+FV}{(1+r)^{N−\frac{t}{T}}}$$
<div style="text-align: right"> Equation 2 </div>

The numerator is the coupon payments and $FV$ of par value at maturity discounted back to time 0 at YTM, and then multiply by the time to get the cashflow (which is in terms of periods, not years).

Equation 3 combines Equations 1 and 2 to reveal an important aspect of the Macaulay duration: **Macaulay duration is a weighted average of the time to receipt of the bond’s promised payments, where the weights are the shares of the full price that correspond to each of the bond’s promised future payments.**

$$MacDur=(1−\frac{t}{T})\times\frac{\frac{PMT}{(1+r)^{1−\frac{t}{T}}}}{PV^{Full}}+(2−\frac{t}{T})\times\frac{\frac{PMT}{(1+r)^{2−\frac{t}{T}}}}{PV^{Full}}+⋯+(N−\frac{t}{T})\times\frac{\frac{PMT+FV}{(1+r)^{N−\frac{t}{T}}}}{PV^{Full}}$$
<div style="text-align: right"> Equation 3 </div>

The times to receipt of cash flow measured in terms of time periods are $1 – t/T$, $2 – t /T$, …, $N – t/T$. The weights are the present values of the cash flows divided by the full price. Therefore, Macaulay duration is measured in terms of time periods.

**If we are calculating duration at issue then $t/T = 0$** For example, a 5-year, 6% semiannual coupon bond yielding 4% has 10 periods. In per period terms, the coupon rate and YTM are 3% and 2%, respectively.

#### Macaulay Duration: Another Approach
Another approach to calculating the Macaulay duration is to use a closed-form equation derived using calculus and algebra. Equation 4 is a general closed-form formula for determining the Macaulay duration of a fixed-rate bond, where c is the coupon rate per period (PMT/FV).

The Macaulay duration of a fixed-rate bond can be calculated using the following formula:
$$MacDur=\Big(\frac{1+r}{r}−\frac{1+r+[N×(c−r)]}{c×[(1+r)^N−1]+r}\Big)−\frac{t}{T}$$
<div style="text-align: right"> Equation 4 </div>
    
where
$t$ = number of days since the last coupon payment

$T$ = number of days in each coupon period

$c$ = coupon rate (per period)

$r$ = YTM (per period)

$N$ = number of periods until maturity

$t/T$ = the fraction of the current period that has gone by

Note that $c$ and $r$ are stated as the `rates per period`. **Its output is the Macaulay duration in terms of periods. It is converted to annual duration by dividing by the number of periods in the year.**

### Modified Dration
`Modified duration` is the Macaulay duration divided by one plus the yield per period. 
$$ModDur=\frac{MacDur}{1+r}$$
<div style="text-align: right"> Equation 5 </div>

Although modified duration might seem to be just a Macaulay duration with minor adjustments, it has an important application in `risk measurement`: **Modified duration provides an estimate of the percentage change in the full bond price (flat price plus accrued interest), for a 1% decrease in the YTM**:

$$\%ΔPV^{Full}≈−AnnModDur×ΔYield$$
<div style="text-align: right"> Equation 6 </div>

*Notice the formula is using the annual modified duration and the annual yield-to-maturity. The minus sign indicates that bond prices and yields-to-maturity move inversely.*

It is important to note that estimated price changes based on modified duration are `linear approximations`. As shown in the graph below, modified duration estimates will **understate** the sensitivity of a bond's price to changes in its YTM.

### Approximate Modified and Macaulay Duration
The modified duration statistic for a fixed-rate bond is easily obtained if the Macaulay duration is already known. An alternative approach is to approximate modified duration directly. Equation 7 is the approximation formula for annual modified duration.

The modified duration can also be directly calculated by estimating the slope of the tangent to the price-yield curve. The original price ($P_0$) can be recalculated by shifting the yield up and down by $ΔYield$.

$$ApproxModDur=\frac{(PV_−)−(PV_+)}{(2)(ΔYield)(PV_0)}$$
<div style="text-align: right"> Equation 7 </div>

The objective of the approximation is to estimate the slope of the line tangent to the price–yield curve. The slope of the tangent and the approximated slope are shown in the below Approximate Modified Duration graph. 

To estimate the slope, 
1. the yield-to-maturity is changed up and down by the same amount—the ΔYield. 
2. Then the bond prices given the new yields-to-maturity are calculated. (The price when the yield is increased is denoted $PV_+$. The price when the yield-to-maturity is reduced is denoted $PV_−$. The original price is $PV_0$.) These prices are the `full prices`, including accrued interest.
3. The slope of the line based on $PV_+$ and $PV_−$ is the approximation for the slope of the line tangent to the price–yield curve. 

Equation 7 can be transformed into a more intuitive formula, i.e. the percentage change of the price over the change in yields:
$$ApproxModDur=\frac{\frac{(PV_−)}{(PV_0)}−\frac{(PV_+)}{(PV_0)}}{(2)\times(ΔYield)}$$

In fact, as ΔYield approaches zero, the approximation approaches AnnModDur.

Although duration can also be estimated quite accurately using the `basic bond-pricing equation` and a financial calculator. The Macaulay duration can be approximated as well—the approximate modified duration multiplied by one plus the **yield per period**.

$$ApproxMacDur=ApproxModDur×(1+r)$$
<div style="text-align: right"> Equation 8 </div>

The approximation formulas produce results for annualized modified and Macaulay durations. The frequency of coupon payments and the periodicity of the yield-to-maturity are included in the bond price calculations.

### Typical Curve Duration: Effective Duration
Another approach to assess the interest rate risk of a bond is **to estimate the percentage change in price given a change in a benchmark yield curve**. The `effective duration` of a bond is the sensitivity of the bond’s price to a change in a benchmark yield curve. The formula to calculate effective duration (EffDur) is Equation 9.

$$EffDur=\frac{(PV_−)−(PV_+)}{2×(ΔCurve)×(PV_0)}$$
<div style="text-align: right"> Equation 9 </div>

Effective duration is a curve duration statistic in that *it measures interest rate risk in terms of a parallel shift in the benchmark yield curve ($ΔCurve$)*.

Effective duration is essential to the measurement of the interest rate risk of a complex bond, such as a bond that contains an embedded call option. The duration of a callable bond is not the sensitivity of the bond price to a change in the yield-to-worst (i.e., the lowest of the yield-to-maturity, yield-to-first-call, yield-to-second-call, and so forth). The problem is that future cash flows are uncertain because they are contingent on future interest rates. The issuer’s decision to call the bond depends on the ability to refinance the debt at a lower cost of funds. In brief, a callable bond does not have a well-defined internal rate of return (yield-to-maturity). Therefore, yield duration statistics, such as modified and Macaulay durations, do not apply; effective duration is the appropriate duration measure.

The option-pricing model inputs include 
1. the length of the call protection period, 
2. the schedule of call prices and call dates, 
3. an assumption about credit spreads over benchmark yields (which includes any liquidity spread as well), 
4. an assumption about future interest rate volatility, and 
5. the level of market interest rates (e.g., the government par curve). 

In practice, a callable bond issuer might be able to exercise the call option and obtain a lower cost of funds if 
- (1) benchmark yields fall and the credit spread over the benchmark is unchanged or 
- (2) benchmark yields are unchanged and the credit spread is reduced (e.g., because of an upgrade in the issuer’s rating). 

A pricing model can be used to determine a “`credit duration`” statistic—that is, the sensitivity of the bond price to a change in the `credit spread`. 

- On a traditional fixed-rate bond, modified duration estimates the **percentage price change** for a change in the benchmark yield and/or the credit spread. 
- For bonds that do not have a well-defined internal rate of return because the future cash flows are not fixed—(for instance, callable bonds and floating-rate notes)—pricing models are used to produce different statistics for changes in benchmark interest rates and for changes in credit risk.
- Another fixed-income security for which yield duration statistics, such as modified and Macaulay durations, are not relevant is a mortgage-backed bond.

The effective duration may also be referred to as `OAS duration` because it is based on the option-pricing model that is also used to calculate the `option-adjusted spread`.

##### The Difference between Modified Duration and Effective Duration
The modified duration and effective duration on an option-free bond are identical only in the rare circumstance of an absolutely flat yield curve.

The difference is less when:
- the Yield curve is flatter
- Time-to-maturity is shorter
- Bond is priced closer to par
- Modified and effective durations are identical for option-free bonds only if the yield curve is flat.

### Typical Curve Duration: Key Rate Duration 
`Key rate duration` (a.k.a `partial duration`) measures the sensitivity of a bond's price to changes in the benchmark yield curve at a *specific `maturity segment`*. 

Key rate duration measures allow us to isolate the impact of changes in specific rates and model the impact of non-parallel shifts in the yield curve. The potential for non-parallel shifts is called `shaping risk`. (Capturing shape changes in the yield curve, which accounts for non-parallel shifts in the yield curve).

*A bond's effective duration will be equal to the sum of its key rate durations.*

Example

The price of a 10-year bond with an effective duration of 7.8 will increase by approximately 7.8% if there is a 1% decrease in all benchmark rates. The same bond can have a 5-year key rate duration of 2.1, which tells us to expect a 2.1% change in the bond's price if there is a 1% change in the 5-year rate and all other rates remain at their current levels.

### Duration of a Bond Portfolio
The duration of a bond portfolio can be thought of as 
1. the duration of the aggregate cash flows or 
2. the weighted average of the individual bond durations. 

There are two ways to calculate the duration of a bond portfolio: 
- (1) the weighted average of time to receipt of the aggregate cash flows, and 
- (2) the weighted average of the individual bond durations that comprise the portfolio. The shares of overall portfolio `market value` are the weights. 

The first method is theoretically correct, but the second method is more practical. The two methods give the same result if the yield curve is flat. They are close if the bonds in the portfolio have similar yields-to-maturity. The second method works for floating-rate bonds and bonds with embedded options while the first method does not.

A `cash flow yield` is the internal rate of return on a series of cash flows, usually used on a complex security such as a mortgage-backed bond (using projected cash flows based on a model of prepayments as a result of refinancing) or a portfolio of fixed-rate bonds.

*Note: use CF function in calculator then find the IRR for cash flow yield.*

#### Limitations of the First Approach
- First, the `cash flow yield` is not commonly calculated for bond portfolios. 
- Second, the amount and timing of future coupon and principal payments are uncertain if the portfolio contains callable or putable bonds or floating-rate notes. 
- Third, interest rate risk is usually expressed as a change in benchmark interest rates, not as a change in the cash flow yield. 
- Fourth, the change in the cash flow yield is not necessarily the same amount as the change in the yields-to-maturity on the individual bonds. (For instance, if the yields-to-maturity on the two zero-coupon bonds in this portfolio both increase or decrease by 10 bps, the cash flow yield increases or decreases by only 9.52 bps.)

#### Advantages and Limitations of the Second Approach
- An advantage of the second approach is that callable bonds, putable bonds, and floating-rate notes can be included in the weighted average using the effective durations for these securities. 
- The main advantage to the second approach is that it is easily used as a measure of interest rate risk. For instance, if the yields-to-maturity on the bonds in the portfolio increase by 100 bps, the estimated drop in the portfolio value is 14.3725%. 
- However, this advantage also indicates a limitation: This measure of portfolio duration implicitly assumes a `parallel shift` in the yield curve. A parallel yield curve shift implies that all rates change by the same amount in the same direction. In reality, interest rate changes frequently result in a steeper or flatter yield curve. (refer to upcoming `yield volatility` topic.)

#### Money Duration and the Price Value of a Basis Point
`Money duration` is a measure of the price change in currency units rather than percentage terms. This is called the dollar duration in the United States. For a given change in the annual yield-to-maturity ($ΔYield$), modified duration estimates the percentage price change and money duration estimates the change in currency units.

$$MoneyDur=AnnModDur\times PV_{Full}$$
<div style="text-align: right"> Equation 10 </div>
$$ΔPV_{Full}≈−MoneyDur\times ΔYield$$
<div style="text-align: right"> Equation 11 </div>

The `price value of a basis point (PVBP)` is another version of money duration. It is the absolute value of the price change for a 1 basis point change in yield.

$$PVBP=\frac{(PV_−)−(PV_+)}{2}$$
<div style="text-align: right"> Equation 12 </div>

where $PV_−$ and $PV_+$ are calculated by shifting the YTM up and down one basis point.

PVBP is simply money duration for a 1 basis point change in yield.

## Bond Convexity
The convexity statistic for the bond is used to improve the estimate of the percentage price change provided by modified duration alone.  Duration (in particular, money duration) estimates the change in the bond price along the straight line that is tangent to the curved line. 

Option-free and putable bonds always exhibit positive convexity. Callable bonds have negative convexity at lower interest rates because the price does not increase as much as would be expected using a duration-based linear estimate.

The convexity-adjusted estimate of the percentage change in the bond’s full price.5

$$\%ΔPV^{Full}≈(−AnnModDur×ΔYield)+[\frac{1}{2}×AnnConvexity×(ΔYield)^2]$$
<div style="text-align: right"> Equation 13 </div>

The first bracketed expression, the “first-order” effect, is the same as Equation 6 $\%ΔPV^{Full}$. 

The second bracketed expression, the “second-order” effect, is the `convexity adjustment`.
- This additional term is a positive amount on a traditional (option-free) fixed-rate bond for either an increase or decrease in the yield. 

Similar to the Macaulay and modified durations, the annual convexity statistic can be calculated in several ways. It can be calculated using tables, such as Exhibits 2 and 3. It also is possible to derive a closed-form equation for the convexity of a fixed-rate bond on and between coupon payment dates using calculus and algebra.6 But like modified duration, convexity can be approximated with accuracy. Equation 14 is the formula for the approximate convexity statistic, ApproxCon.

A bond's convexity can be approximated using the following formula:
$$ApproxCon=\frac{(PV_−)+(PV_+)−[2×(PV_0)]}{(ΔYield)^2×(PV_0)}$$
<div style="text-align: right"> Equation 14 </div>

These are the full prices, including accrued interest, for the bond.

The Macaulay duration of a zero-coupon bond is N – t/T in terms of periods to maturity. The exact convexity statistic of a zero-coupon bond, also in terms of periods, is calculated with Equation 15.
$$Convexity (of\ a\ zero\ coupon\ bond)=\frac{[N−(t/T)]×[N+1−(t/T)]}{(1+r)^2}$$
<div style="text-align: right"> Equation 15 </div>

N is the number of periods to maturity as of the beginning of the current period, t/T is the fraction of the period that has gone by, and r is the yield-to-maturity per period.

**As with the other statistics, convexity is annualized** in practice and for use in the convexity adjustment in Equation 13. It is divided by the periodicity squared. (For example, on a semiannual bond basis, meaning a periodicity of 2.)

In general, the differences are heightened when the benchmark yield curve is not flat, when the bond has a long time-to-maturity, and the bond is priced at a significant discount or premium.

## Investment Horizon, Macauley Duration and Interest Rate Risk
### Interest Rate Risk and the Investment Horizon
When an investor has a short-term horizon, duration (and convexity) are used to estimate the change in the bond price. In this case, yield volatility matters. In particular, bonds with varying times-to-maturity have different degrees of yield volatility. When an investor has a long-term horizon, the interaction between coupon reinvestment risk and market price risk matters. 
#### Yield Volatility: Short-term
In a short-term horizon, the accrued interest does not change, thus the interest rate risk (from YTM) mainly concerns the flat price. 

Bond duration is the primary measure of risk arising from a change in the yield-to-maturity. Convexity is the secondary risk measure.

The importance of yield volatility in measuring interest rate risk is that bond price changes are products of two factors: (1) the impact per basis-point change in the yield-to-maturity and (2) the number of basis points in the yield-to-maturity change. The first factor is duration or the combination of duration and convexity, and the second factor is the yield volatility. (Refer to example 15)

$$\%ΔPV^{Full}≈(−AnnModDur×ΔYield)+[\frac{1}{2}×AnnConvexity×(ΔYield)^2]$$
#### Investment Horizon, Macaulay Duration, and Interest Rate Risk: Long-term 
The investor faces coupon reinvestment risk as well as market price risk if the bond needs to be sold prior to maturity.

Macaulay duration represents the investment horizon that is immune to interest rate changes. If an investor's expected holding period is equal to the Macaulay duration, coupon reinvestment risk and market price risk will exactly offset each other when interest rates change in a one-time parallel fashion. A rise in rates will lead to an increase in reinvestment income that is offset by a market loss on the sale. A drop in rates will lead to a decrease in reinvestment income that is offset by a market gain on the sale.

An imbalance between investment horizon and Macaulay duration is called a `duration gap`, which is calculated as follows:
$$Duration\ gap=MacDur−Investment\ horizon$$

An investor with a short time horizon relative to the Macaulay duration has a positive duration gap. Changes in interest rates will have more influence on bond prices than returns from reinvesting coupons, so the investor will suffer losses if interest rates rise.

By contrast, a negative duration gap means that market price risk is dominated by coupon reinvestment risk, and investors are exposed to the risk of falling interest rates.

The benchmark yield curve reflects real interest rates and inflation expectations. Yields on corporate bonds include a spread above benchmark rates to reflect the premiums that investors require as compensation for accepting default risk and liquidity risk. Measures of credit duration and liquidity duration can be used to estimate a bond's sensitivity to changes in these premiums.

# Fundamentals of Credit Analysis

## Introduction to Credit Risk
`Credit risk` is the risk of loss from the borrower defaulting on interest or principal payments. The two components of credit risk are:
1. `Default risk` (or `default probability`): The probability the borrower defaults.
2. `Loss severity`: The loss given a default occurs.

The `expected loss` is the product of the default probability (or default risk) and the loss severity.

$$Expected\ Loss = Default\ Probability \times Loss\ Severity$$

$$Loss\ Severity = 1 – Recovery\ Rate$$
Where the recovery rate is the percentage of the principal amount recovered in the event of default.

`Yield spreads`, expressed in basis points, widen based on two primary factors: 
- (1) a decline in an issuer’s creditworthiness, sometimes referred to as credit migration or downgrade risk, and 
- (2) an increase in market liquidity risk. 

`Spread risk` is the risk that investors will require higher spreads on corporate bonds relative to the government yield curve. Two factors that contribute to spread risk are 
1. `Credit migration risk` (or `downgrade risk`): the possibility that a borrower's credit rating will be lowered. It is also known as downgrade risk.
2. `Market liquidity risk`: the threat of investors being unable to trade bonds without having to sell at a large discount or pay a significant premium. Liquidity can decline rapidly in times of financial stress. To compensate investors for the risk that there may not be sufficient market liquidity for them to buy or sell bonds in the quantity they desire, the spread or yield premium on corporate bonds includes a market liquidity component, in addition to a credit risk component.

    *Unlike stocks, which trade on exchanges, most markets bonds trade primarily over the counter, through broker–dealers trading for their own accounts. Their ability and willingness to make markets, as reflected in the bid–ask spread, is an important determinant of market liquidity risk.*

## Capital Structure, Seniority Ranking, and Recovery Rates
### Capital Structure
`Capital structure` refers to a company's debt and equity mix. Some companies have very complicated capital structures that have many different rankings of debt.
### Seniority Ranking
Not all debt obligations from the same issuer will have the same seniority ranking. More senior bonds will have a higher priority of payments in the event of a liquidation. A company with multiple bond issues can have the following seniority ranking:
- First Lien Loan - Senior Secured
- Second Lien Loan - Secured
- Senior Unsecured
- Senior Subordinated
- Subordinated
- Junior Subordinated

Secured debt has a direct claim on specified assets. The specified assets could be property (as in first mortgage debt) or other assets such as equipment and patents. Second and third liens have a lower priority. Second lien debt has higher seniority than a senior unsecured debt because the former is secured while the latter is not.

Unsecured debt (or debentures) only have a general claim on the issuer's assets. This is the most common type of corporate bond. Some issues can be subordinated.
### Recovery Rates
All creditors at the same level regardless of maturity have the same claim in bankruptcy. This is referred to as bonds ranking `pari passu` (“on an equal footing”). Defaulted debt has some value based on recovery expectations.

Many statistical studies have been conducted on recovery rates. Higher priority claims have significantly greater recovery rates. The recovery rates also vary widely by industry. The rates are typically lower if the default occurs at the bottom of a `credit cycle`. Understanding recovery rates is very important when estimating the expected loss.

Theoretically, secured creditors get the value of the specified property before any other claim, unsecured creditors are paid in full before equity holders, and senior creditors have priority over junior creditors. In practice, the priority of claims is not always absolute.

*(Note: All impaired classes get to vote on the reorganization plan. Negotiation and compromise are often preferable to incurring huge legal and accounting fees in a protracted bankruptcy process that would otherwise reduce the value of the estate for all claimants. This process may allow junior creditors (e.g., subordinated bondholders) to recover some value even though more senior creditors do not get paid in full.)*

All classes of claimants get to vote on the reorganization details. This often results in negotiation, which can take a lot of time and money. To speed up the negotiation process, `concessions` are often made to equity holders and junior creditors.

In bankruptcy proceedings, `reorganization` is more common in the United States, while `liquidation` is preferred in other jurisdictions like the United Kingdom. Liquidation is generally more favorable for senior creditors.

## Ratings Agencies, Credit Ratings, and Their Role in the Debt Markets
The major credit rating agencies are Moody's, Standard & Poor's, and Fitch. Issuers and investors often require these ratings. Investors like the independent analysis that makes it easy to compare various issues. Regulators rely on ratings to help manage the huge debt markets.

The rating agencies were criticized for their role in the financial crisis of 2008-2009. Rules were passed to promote transparency, objectivity, and competition.

- Bonds rated triple-A or double-A are high-quality grades. 
- Bonds rated single-A or triple-B are medium grade. 
- The D rating is reserved for securities that are already in default in S&P’s and Fitch’s scales. For Moody’s, bonds rated C are likely, but not necessarily, in default. 

Agencies also often give an outlook of positive, stable, or negative. Also “On Review for Downgrade” or “On CreditWatch for an Upgrade.”
### Issuer vs. Issue Ratings
- An `issuer rating` addresses the borrower’s overall creditworthiness and usually only applies to senior unsecured debt. 
- `Issue rating` takes into account the ranking in the capital structure. 

Terminology used to distinguish between issuer and issue ratings includes `corporate family rating (CFR)` and `corporate credit rating (CCR)` or `issuer credit rating` and `issue credit rating`. 
#### Notching
The issuer credit rating usually applies to its senior unsecured debt (related: `Cross-default provisions`). `Notching` enables specific issues to get different ratings. Notching is the ratings adjustment methodology where specific issues from the same borrower may be assigned different credit ratings. Factors lead to the need of notching include: 
- Priority of payment in the event of a default (e.g., secured versus senior unsecured versus subordinated) 
- Potential loss severity in the event of default
- `Structural subordination` can arise when a corporation with a holding company structure has debt at both its parent holding company and operating subsidiaries. Debt at the operating subsidiaries will get serviced by the cash flow and assets of the subsidiaries before funds can be passed (“upstreamed”) to the holding company to service debt at that level.

As a general rule, **the higher the senior unsecured rating, the smaller the notching adjustment will be.** The reason behind this is that the higher the rating, the lower the perceived risk of default, so the need to “notch” the rating to capture the potential difference in loss severity is greatly reduced. (For lower-rated credits, however, the risk of default is greater and thus the potential difference in loss from a lower (or higher) priority ranking is a bigger consideration in assessing an issue’s credit riskiness. Thus, the rating agencies will typically apply larger rating adjustments.)
### Risks in Relying on Agency Ratings
Historically, the ratings have proved pretty accurate regarding default risk. High-rated bonds have very low default rates. However, the following limitations of ratings should be considered.
- Credit ratings are dynamic (i.e., bond investors should not assume an issuer’s credit rating will remain the same from time of purchase through the entire holding period.).
- Ratings tend to lag the market pricing of risk. The market reacts daily to news of credit risk while rating agencies are slow to make adjustments. The market focuses on the expected loss while rating agencies focus on the probability of default.
- Rating agencies are not infallible. For example, the agencies missed the risk in subprime-backed mortgage securities.
- Event risks are difficult to capture in ratings. `Litigation` of tobacco companies and tsunamis in Japan can affect the creditworthiness of companies, but they are tough to predict. Leveraged transactions, such as debt-financed acquisitions and large stock buybacks (share repurchases), are often difficult to anticipate and thus to capture in credit ratings.

## Traditional Credit Analysis: Corporate Debt Securities
#### Credit Analysis vs. Equity Analysis: Similarities and Differences
A company's managers have legal obligations to creditors, but they work for shareholders who want to see growth and justifiable risks taken. By contrast, lenders are primarily concerned about a company's ability to generate stable cash flows to service its debts. Given these differing perspectives, equity analysts place far more emphasis on a company's `upside potential` and `income statement`, while credit analysts look at `downside risk` and the `balance sheet`.
### The Four C's of Credit Analysis: Capacity 
`Capacity` is the ability of the borrower to make timely payments. The analyst must perform **industry and company analysis**.

Michael Porter identified the five major forces of industry competition to break down the industry structure.
1. Power of suppliers<br>
Credit risk tends to be higher industries where firms have low bargaining power compared to a relatively small number of suppliers.

2. Power of buyers/customers<br>
Industries with very few customers have more credit risk because all the negotiating power is with the buyers.

3. Barriers to entry<br>
Industries with high entry barriers have less credit risk. The entry barriers can be from the capital, technology, or regulation.

4. Substitution risk<br>
There is less credit risk if there are no good substitute products or services. Over time technology often creates substitutes.

5. Level of competition<br>
Industries with greater competition have more credit risk.

**Industry Fundamentals**
- Sensitivity to macroeconomic factors
- Growth prospects
- Profitability <br>
Cyclical industries (like the auto industry) are riskier because they are more sensitive to broad economic performance. Industries with slow growth tend to consolidate through mergers and acquisitions that can hurt bondholders. Many industry statistics are available to the analyst.

**Company Fundamentals**<br>
The fundamentals of the company should also be analyzed. This includes considering the competitive position, track record, management's strategy, and ratios. The competitive position of the company within the industry should be determined. Consider its current and trending market share and relative cost structure. The track record and operating history should be analyzed over several years. Management strategy and execution should be evaluated for plausibility and consistency.
#### Capacity Ratios and Ratio Analysis
Financial ratios from the income statement, balance sheet, and cash flow statement should be analyzed. The key credit ratios can be placed into three groups.
##### Profitability and Cash Flow
Profitability and cash flow are needed to service debt. Usually, the focus is on `operating income`, which is earnings before interest and taxes (`EBIT`). There are several measures of cash flow.
- Earnings before interest, taxes, depreciation, and amortization (`EBITDA`) adds back some non-cash items to earnings. It ignores some cash expenses such as capital expenditures.
- Funds from operations (`FFO`) is net income from continuing operations plus depreciation, amortization, deferred income taxes, and other non-cash items.
- Free cash flow before dividends (`FCF before dividends`) is net income plus depreciation and amortization minus capital expenditures minus increase in non-cash working capital. Companies with negative free cash flow must seek additional financing.
- Free cash flow after dividends (`FCF after dividends`) simply subtracts the dividends from the above calculation.

##### Leverage
- The debt to capital ratio uses capital equal to total debt plus shareholders’ equity. It represents the percentage of capital financed with debt.
- The debt to EBITDA ratio is used by rating agencies and banks in loan covenants. It can be volatile for cyclical companies.
- The FFO to debt ratio is liked by credit rating agencies.
- FCF after dividends/debt.

##### Coverage 
Coverage ratios measure the issuer's ability to "cover" its interest payments. The two common ratios are 
- EBITDA/interest expense and 
- EBIT/interest expense<br>
The difference between the two is that EBITDA puts back depreciation and amortization expenses. This will result in a higher figure, making EBIT/interest expense the more conservative measure. However, EBITDA/interest expense ratio is more frequently used in practice.

##### Issuer Liquidity
Liquidity is also important to consider. Companies with strong liquidity have less credit risk. The liquidity can be assessed by looking at:
- Cash on the balance sheet, 
- Net working capital, 
- Operating cash flow, 
- Committed bank lines, and 
- Imminent debt and capital expenditures (Debt coming due and committed capital expenditures in the next one to two years).

### The Four C's of Credit Analysis: Collateral, Covenants, and Character
**Collateral** Analysis of collateral is generally more important for weaker companies. Collateral value is relevant for loss severity when a company does default. The value and quality of assets should be considered. Companies that trade below book value often have lower-quality assets. Some companies like software companies rely on the intellectual property rather than real assets.

**Covenants** Covenants protect creditors and give management flexibility. Affirmative covenants are obligations and restrictive covenants are limitations. Corporate bonds describe the covenants in the prospectus. Strong covenants can protect the bondholders from management making decisions that only enrich the shareholders.

**Character** Assessment of character is difficult because most companies are publicly owned rather than being owned by management. The management's strategy, track record, use of aggressive accounting policies, history of fraud, and previous poor treatment of bondholders are all important.

## Impact of Credit Spread on Risk and Return (Credit Risk vs. Return)
### Yields and Spreads
Taking on more credit risk offers greater potential returns but also more risk. The quoted yields on non-investment grade bonds are significantly higher, but the realized yields depends on the actual default and downgrade experience.

$$Yield\ on\ Corporate\ Bond=Real\ Risk-free\ Interest\ Rate+Expected\ Inflation\ Rate+Maturity\ Premium+Liquidity\ Premium+Credit\ Spread$$

Investors often focus on the `yield spread` relative to a default-free bond.
$$Yield\ Spread=Liquidity\ Premium+Credit\ Spread$$

Corporate bond spreads are affected by several factors. Credit spreads will widen if the credit cycle deteriorates, economy conditions worsen, financial markets weaken, broker-dealers' reduce capital commitments, or market supply increases.
### The Price Impact of Spread Changes
Investors in investment-grade bonds focus more on `spread risk` (the effect on prices and returns from changes in spreads) rather than default risk because the probability of default is very low. The bond's return from spread changes is driven by its modified duration (price sensitivity) and size of spread change.

$$Price\ Impact≈−(MDur×ΔSpread)+12(Cvx)(ΔSpread)^2$$
 
where $MDur$ is the modified duration and $Cvx$ is the convexity.

Longer-duration bonds have greater spread sensitivity. Also, there is more uncertainty regarding future creditworthiness for longer-maturity bonds. Bid-ask spreads are also normally greater for longer-duration bonds. All of this leads to spread curves with a positive slope (i.e., spread increases with maturity).

## Special Considerations (High-Yield, Sovereign, and Municipal Credit Analysis) 
### High-Yield Credit Analysis
`High yield` (or junk) bonds are rated below Baa3/BBB-. Companies are rated below investment grade for reasons including high leverage, weak operating history, limited free cash flow, or poor management.

High-yield analysis is normally more involved than investment-grade analysis.

**Liquidity** 
Liquidity is critical for high-yield companies. Investment-grade companies can easily refinance maturing debt. This is not true for high-yield companies, who do not have access to commercial paper and are often privately held, thus limiting equity financing.

Sources of liquidity, from strongest to weakest, are: cash, working capital, operating cash flow, bank credit, equity issuance, and asset sales. The amount of liquidity sources should be compared to the upcoming debt maturities. Insufficient liquidity is the number one reason for default.

**Financial projections**
Since there is less certainty for high-yield companies, it is important to project several financial forecasts, focusing on stress scenarios.

**Debt structure**
High-yield companies often have many layers of debt with various seniorities. Lower rankings in the debt structure translate to lower credit ratings.

Companies with a lot of secured debt are said to be "`top-heavy`". This leads to less capacity to take on bank debt.

**Corporate structure**
Many companies have holding company structures. The analyst must carefully consider parent and subsidiary debt and cash movements between them. Parent companies usually rely on dividends from subsidiaries. The parent debt is usually subordinated to the subsidiary debt. Some have multiple parent and subsidiary arrangements.

**Covenant Analysis**
Key covenants for high-yield issuers could include the following:
- Change of control put: In a change of control put, the buyer can force the issuer to buy back the debt if an acquisition occurs. This protects the borrower from weak corporate acquirers.
- Restricted payments: Restricted payments covenants place limits on cash paid to shareholders. The limit is usually based on cash flow and debt outstanding.
- Limitations on liens: Limitations on lien covenants restrict the amount of secured debt. This protects unsecured creditors.
- Restricted versus unrestricted subsidiaries: Restricted subsidiaries offer more guarantees on the holding company debt. However, some indentures allow the corporation the ability to reclassify at will.
- Bank covenants: Bank covenants are typically very restrictive and could block future loans.
- Equity-like analysis approach: High-yield bonds are sometimes considered hybrid security between investment-grade bonds and equity. They are more correlated with equity movements than interest rates. Enterprise value (market capitalization plus debt minus excess cash) can be used to determine leverage potential at the company.

### Sovereign Debt
Governments issue debt to finance general operations. Bonds issued by governments like the United States are often considered risk-free. Weak countries must issue bonds in foreign currency that is more stable. That eliminates the option to repay debt with freshly printed money that devalues the currency.

Sovereign defaults can be caused by events such as war and political change. Default risk for developed countries escalated in 2009. Ultimately, sovereign credit risk depends on the government's ability and willingness to pay.

The political and economic profile should be analyzed. One must consider the effectiveness, stability, and predictability of policymaking. This includes handling of the past political and financial crisis, checks and balances, and the absence of corruption. There must also be a commitment to honor debts. A stable economic structure with growth is preferred.

The flexibility and performance profile should also be analyzed. One should focus on the external liquidity and international investment position and the fiscal performance, flexibility, and debt burden. Monetary flexibility is critical, including the credibility and effectiveness of the monetary policy.
### Non-Sovereign Government Debt
Municipal debt includes issues by state, provincial, and local governments. The United States municipal debt market is the largest. Most municipal bonds are tax-exempt. The historical default rate has been very low. Most are general obligation or revenue bonds.

General obligation bonds are unsecured. The credit analysis is similar to sovereign debt analysis, but municipalities must balance their operating budget and can't use monetary policy. The analysis considers employment, income, debt, and tax base. It is riskier to depend on only one or two types of tax.

Revenue bonds are issued to finance a specific project and have more risk than general obligation bonds. The credit analysis must consider the project. The analysis is similar to that of a corporate bond.
